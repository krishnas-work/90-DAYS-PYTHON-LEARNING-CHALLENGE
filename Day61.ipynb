```python
# Q1: Count how many characters in a string are letters (ignore spaces and digits)

# sol:
a = "Python 3 is fun"
c = 0

for i in a:
    if i.isalpha():
        c = c + 1

print(c)
```

```python
# Q2: Create a new list containing numbers that are between 10 and 25

# sol:
a = [5, 12, 18, 27, 22, 9]
b = []

for i in a:
    if i > 10:
        if i < 25:
            b.append(i)

print(b)
```

```python
# Q3: Find the largest number in a tuple.

# Sol:
a = (10, 25, 7, 30, 18)

m = a[0]

for i in a:
    if i > m:
        m = i

print(m)
```

```python
# Q4: Create a new set containing elements that are less than or equal to 4

# sol: 
a = {1, 2, 3, 4, 5, 6}
b = set()

for i in a:
    if i <= 4:
        b.add(i)

print(b)
```

```python
# Q5: Create a function that takes a dictionary and returns the sum of all values.

# sol:
def fun(a):
    s = 0

    for i in a:
        s = s + a[i]

    return s

print(fun({"a": 5, "b": 10, "c": 3}))
```
