```python
# Q1: Count how many characters in a string are not vowels.

# sol:
a = "python code"
c = 0

for i in a:
    if i not in "aeiou" and i != " ":
        c = c + 1

print(c)
```

```python
# Q2: Create a new list containing numbers that are multiples of 4.

# sol:
a = [3, 8, 10, 16, 5]
b = []

for i in a:
    if i % 4 == 0:
        b.append(i)

print(b)
```

```python
# Q3: Count how many elements are greater than 10.

# sol:
a = (5, 12, 7, 18, 3)
c = 0

for i in a:
    if i > 10:
        c = c + 1

print(c)
```

```python
# Q4: Create a new set containing elements that are divisible by 2.

# Sol:
a = {1, 2, 3, 4, 5, 6}
b = set()

for i in a:
    if i % 2 == 0:
        b.add(i)

print(b)
```

```python
# Q5: Create a function that takes a dictionary and returns the count of values less than 10.

# sol:
def fun(a):
    c = 0

    for i in a:
        if a[i] < 10:
            c = c + 1

    return c

print(fun({"a": 5, "b": 12, "c": 7, "d": 20}))
```
