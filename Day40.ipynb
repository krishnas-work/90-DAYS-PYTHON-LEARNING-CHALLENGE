```python
# Q1: Find the longest word in a string.

# sol:
a = "python coding is very interesting"
long = ""

for i in a.split():
    if len(i) > len(long):
        long = i

print(long)
```

```python
# Q2: From a list, create a new list containing numbers whose square is greater than 50.

# Sol:
a = [3, 5, 8, 4, 9]
b = []

for i in a:
    if i * i > 50:
        b.append(i)

print(b)
```

```python
# Q3: Find the sum of elements at odd index positions in a tuple.

# sol: 
a = (10, 20, 30, 40, 50)
s = 0

for i in range(len(a)):
    if i % 2 != 0:
        s = s + a[i]

print(s)
```

```python
# Q4: Create a new dictionary containing only items whose keys have length greater than 3.

# sol:
a = {"pen": 10, "book": 20, "pencil": 15, "bag": 25}
b = {}

for i in a:
    if len(i) > 3:
        b[i] = a[i]

print(b)
```

```python
# Q5: Create a function that takes a number and returns the reverse of the number.

# sol:
def reverse(n):
    r = 0

    while n > 0:
        r = r * 10 + (n % 10)
        n = n // 10

    return r

print(reverse(1234))
```
