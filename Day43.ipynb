```python
# Q1: Count how many characters in a string are punctuation marks

# sol:
a = "hello, how are you?"
c = 0

for i in a:
    if i in ".,!?":
        c = c + 1

print(c)
```

```python
# Q2: Create a new list containing numbers whose difference with the first element is greater than 5.

# sol:
a = [10, 12, 18, 7, 20]
b = []

first = a[0]

for i in a:
    if abs(i - first) > 5:
        b.append(i)

print(b)
```

```python
# Q3: Find the second largest element in a tuple.

# sol:
a = (10, 40, 25, 60, 30)

m1 = a[0]
m2 = a[0]

for i in a:
    if i > m1:
        m2 = m1
        m1 = i
    elif i > m2 and i != m1:
        m2 = i

print(m2)
```

```python
# Q4: Create a new set containing elements whose square is an even number.

# sol:
a = {1, 2, 3, 4, 5}
b = set()

for i in a:
    if (i * i) % 2 == 0:
        b.add(i)

print(b)
```

```python
# Q5: Create a function that takes a dictionary and returns the sum of all its values.

# sol:
def total(a):
    s = 0

    for i in a:
        s = s + a[i]

    return s

print(total({"x": 10, "y": 20, "z": 30}))
```
