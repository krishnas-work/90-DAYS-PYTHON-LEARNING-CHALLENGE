```python
# Q1: Count how many characters in a string are uppercase consonants or even digits.

# sol:
a = "PyTHoN246 CodE"
c = 0

for i in a:
    if i.isupper():
        if i not in "AEIOU":
            c = c + 1
    elif i.isdigit():
        if int(i) % 2 == 0:
            c = c + 1

print(c)
```

```python
# Q2: Create a new list containing numbers whose sum of digits is less than the number itself.

# sol: 
a = [12, 25, 39, 44, 51]
b = []

for i in a:
    n = i
    s = 0

    while n > 0:
        s = s + (n % 10)
        n = n // 10

    if s < i:
        b.append(i)

print(b)
```

```python
# Q3: Find the count of elements which are equal to the sum of first and last element.

# sol:
a = (10, 20, 30, 40, 50)

s = a[0] + a[-1]

c = 0
for i in a:
    if i == s:
        c = c + 1

print(c)
```

```python
# Q4: Create a function that takes a dictionary and returns a list of keys whose values are divisible by the sum of their digits.

# sol:
def fun(a):
    b = []

    for i in a:
        n = a[i]
        s = 0

        temp = n
        while temp > 0:
            s = s + (temp % 10)
            temp = temp // 10

        if s != 0 and n % s == 0:
            b.append(i)

    return b

print(fun({"a": 12, "b": 21, "c": 25, "d": 18}))
```

```python
# Q5: Create a new set containing elements that are greater than 5 and even

# sol:
a = {2, 4, 6, 7, 8, 3}
b = set()

for i in a:
    if i > 5 and i % 2 == 0:
        b.add(i)

print(b)
```
