```python
# Q1: Count how many characters in a string are digits OR special characters (not letters and not spaces).

# sol:
a = "Python@123 Code#"
c = 0

for i in a:
    if not i.isalpha() and i != " ":
        c = c + 1

print(c)
```

```python
# Q2: Create a new list containing numbers whose square ends with digit 6.

# sol:
a = [2, 4, 5, 6, 8]
b = []

for i in a:
    if (i * i) % 10 == 6:
        b.append(i)

print(b)
```

```python
# Q3:Count how many elements in a tuple are strictly between the smallest and largest value.

# sol:
a = (10, 25, 5, 40, 18)

mx = a[0]
mn = a[0]

for i in a:
    if i > mx:
        mx = i
    if i < mn:
        mn = i

c = 0
for i in a:
    if i > mn and i < mx:
        c = c + 1

print(c)
```

```python
# Q4: Create a new set containing elements whose square minus the number is greater than 20.

# sol:
a = {4, 5, 6, 7}
b = set()

for i in a:
    if (i * i - i) > 20:
        b.add(i)

print(b)
```

```python
# Q5: Create a function that takes a dictionary andreturns the count of values whose sum of digits is greater than 10.

# sol:
def count_val(a):
    c = 0

    for i in a:
        n = a[i]
        s = 0

        while n > 0:
            s = s + (n % 10)
            n = n // 10

        if s > 10:
            c = c + 1

    return c

print(count_val({"a": 56, "b": 23, "c": 89, "d": 11}))
```
