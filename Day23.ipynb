```python
# Q1: Take a string and:
#     count how many alphabet characters are there
#     (ignore digits and special characters)

# sol:
a = "Py@12th!on"
c = 0

for i in a:
    if i.isalpha():
        c = c + 1

print(c)
```

```python
# Q2: From a list of numbers:
#     store each number and its square in a dictionary

# sol:
a = [2, 3, 4]
d = {}

for i in a:
    d[i] = i * i

print(d)
```

```python
# Q3: Take a string and:
#     create a list of all characters except spaces

# sol:
a = "hi all"
b = []

for i in a:
    if i != " ":
        b.append(i)

print(b)
```

```python
# Q4: From a list of numbers:
#     convert it into a tuple
#     count how many numbers in the tuple are negative

# sol:
a = [3, -2, 5, -7, 4]
t = tuple(a)

c = 0
for i in t:
    if i < 0:
        c = c + 1

print(c)
```

```python
# Q5: From a dictionary:
#     create a list of all values greater than 30

# sol:
a = {"x": 10, "y": 40, "z": 25, "w": 60}
b = []

for i in a:
    if a[i] > 30:
        b.append(a[i])

print(b)
```
