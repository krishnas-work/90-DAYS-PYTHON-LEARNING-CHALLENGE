```python
# Q1: Check.if number of uppercase vowels + digits greater than 6 is greater than number of lowercase consonants.

# sol:
a = "Kartik is my bestiee23"

x = 0
y = 0

for i in a:
    if i.isupper():
        if i in "AEIOU":
            x = x + 1
    elif i.isdigit():
        if int(i) > 6:
            x = x + 1
    elif i.isalpha():
        if i.islower():
            if i not in "aeiou":
                y = y + 1

if x > y:
    print("yes")
else:
    print("no")
```

```python
# Q2: Create a new list of elements where value is greater than (index × 2) and less than (index + 5).

# Sol:
a = [3, 5, 9, 7, 12, 6]
b = []

for i in range(len(a)):
    if a[i] > i * 2:
        if a[i] < i + 5:
            b.append(a[i])

print(b)
```

```python
# Q3: Count elements where it is less than previous but greater than next.

# Sol:
a = (8, 5, 6, 3, 4)
c = 0

for i in range(1, len(a) - 1):
    if a[i] < a[i - 1] and a[i] > a[i + 1]:
        c = c + 1

print(c)
```

```python
# Q4: Create a new set of elements whose (value² + value) is even and less than 50.


# sol:
a = {2, 3, 4, 5, 6}
b = set()

for i in a:
    val = i * i + i

    if val % 2 == 0:
        if val < 50:
            b.add(i)

print(b)
```
