```python
# Q1: Check if number of uppercase letters is greater than number of lowercase vowels.

# sol:
a = "PyThOnAEiou12"

u = 0
v = 0

for i in a:
    if i.isupper():
        u = u + 1
    elif i in "aeiou":
        v = v + 1

if u > v:
    print("yes")
else:
    print("no")
```

```python
# Q2: Create a new list of elements where value is divisible by its index (ignore index 0).

# sol:
a = [3, 4, 6, 8, 10]
b = []

for i in range(1, len(a)):
    if a[i] % i == 0:
        b.append(a[i])

print(b)
```

```python
# Q3: Count elements where difference with next element is greater than 2.

# Sol:
a = (2, 5, 9, 10, 14)
c = 0

for i in range(len(a) - 1):
    if abs(a[i] - a[i + 1]) > 2:
        c = c + 1

print(c)
```

```python
# Q4: Create a new set of elements whose square is even.

# Sol:
a = {2, 3, 4, 5}
b = set()

for i in a:
    if (i * i) % 2 == 0:
        b.add(i)

print(b)
```

```python
# Q5: Create a function that takes a list and returns numbers whose first digit is less than last digit.

# sol:
def fun(a):
    b = []

    for i in a:
        last = i % 10

        temp = i
        while temp >= 10:
            temp = temp // 10

        first = temp

        if first < last:
            b.append(i)

    return b

print(fun([23, 41, 56, 72, 19]))
```
