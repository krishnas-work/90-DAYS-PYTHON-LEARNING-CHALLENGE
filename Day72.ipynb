```python
# Q1: Count how many characters in a string are either lowercase vowels OR digits greater than 3.

# sol:
a = "PyThon245 code8"
c = 0

for i in a:
    if i in "aeiou":
        c = c + 1
    elif i.isdigit():
        if int(i) > 3:
            c = c + 1

print(c)
```

```python
# Q2: Create a new list containing numbers that are less than 6 and not equal to 2

# sol:
a = [1, 2, 5, 7, 3, 8]
b = []

for i in a:
    if i < 6:
        if i != 2:
            b.append(i)

print(b)
```

```python
# Q3: Count how many elements are greater than 5 and even.

# sol:
a = (2, 6, 9, 4, 8, 3)
c = 0

for i in a:
    if i > 5:
        if i % 2 == 0:
            c = c + 1

print(c)
```

```python
# Q4: Create a new set containing elements whose value + 4 is less than 10.

# sol:
a = {1, 2, 3, 5, 6}
b = set()

for i in a:
    if i + 4 < 10:
        b.add(i)

print(b)
```

```python
# Q5: Create a function that takes a dictionary and returns the sum of values divisible by 5.

# sol:
def fun(a):
    s = 0

    for i in a:
        if a[i] % 5 == 0:
            s = s + a[i]

    return s

print(fun({"a": 5, "b": 12, "c": 10, "d": 7}))
```
