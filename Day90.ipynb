```python
# Q1:  Check if
#      (lowercase consonants + digits less than 5)
#      is greater than
#      (uppercase vowels + digits greater than 7)

# sol:
a = "I have completed my 90Days python learning challenge"

x = 0
y = 0

for i in a:
    if i.isalpha():
        if i.islower() and i not in "aeiou":
            x = x + 1
        elif i.isupper() and i in "AEIOU":
            y = y + 1
    elif i.isdigit():
        if int(i) < 5:
            x = x + 1
        elif int(i) > 7:
            y = y + 1

if x > y:
    print("yes")
else:
    print("no")
```

```python
# Q2: Create a new list of elements where:
#     value is greater than (index + 2)
#     value is less than (index × 4 + 6)
#     value is odd


# sol:
a = [3, 7, 10, 5, 13, 8, 17]
b = []

for i in range(len(a)):
    if a[i] > i + 2:
        if a[i] < i * 4 + 6:
            if a[i] % 2 != 0:
                b.append(a[i])

print(b)
```

```python
# Q3: Count elements where:
#     greater than previous
#     difference with next is greater than 3


# sol:
a = (2, 6, 5, 11, 7, 15)
c = 0

for i in range(1, len(a) - 1):
    if a[i] > a[i - 1]:
        if (a[i + 1] - a[i]) > 3:
            c = c + 1

print(c)
```

```python
# Q4: Create a new set of elements whose:
#     (value² + value) is divisible by 3
#     result is less than 50
#     value is even


# sol:
a = {2, 3, 4, 5, 6, 7}
b = set()

for i in a:
    val = i * i + i

    if val % 3 == 0:
        if val < 50:
            if i % 2 == 0:
                b.add(i)

print(b)
```

```python
# Q5: Create a function that:
#     takes a list
#     returns numbers where
#     sum of digits at even positions > sum at odd positions


# sol:
def fun(a):
    b = []

    for i in a:
        digits = []
        temp = i

        while temp > 0:
            digits.append(temp % 10)
            temp = temp // 10

        digits.reverse()

        s1 = 0
        s2 = 0

        for j in range(len(digits)):
            if j % 2 == 0:
                s1 = s1 + digits[j]
            else:
                s2 = s2 + digits[j]

        if s1 > s2:
            b.append(i)

    return b

print(fun([1234, 5612, 9870, 4523]))
```
