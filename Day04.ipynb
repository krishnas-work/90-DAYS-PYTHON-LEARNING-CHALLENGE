```python
# Q2: Take a character as input and check whether it is a vowel OR uppercase letter.
# sol: 
x=input("enter a char.")
l="a e i o u"
if x in l or x.isupper():
    print("it is  vowel OR uppercase letter")
else:
    print(" not a vowel OR uppercase letter")
```
