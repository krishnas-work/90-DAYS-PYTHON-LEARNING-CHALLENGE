```python
# Q1: Count how many characters in a string are same as the last character.

# sol:
a = "level"
c = 0

l = a[-1]

for i in a:
    if i == l:
        c = c + 1

print(c)
```

```python
# Q2:  Create a new list by adding 2 to each number.

# sol:
a = [1, 2, 3, 4]
b = []

for i in a:
    b.append(i + 2)

print(b)
```

```python
# Q3: Find the sum of all elements.

# sol:
a = (5, 10, 15, 20)
s = 0

for i in a:
    s = s + i

print(s)
```

```python
# Q4: Create a new set containing elements that are less than 5.

# sol:
a = {2, 6, 3, 8, 1}
b = set()

for i in a:
    if i < 5:
        b.add(i)

print(b)
```

```python
# Q5: Create a function that takes a dictionary and returns the maximum value.

# sol:
def fun(a):
    m = list(a.values())[0]

    for i in a:
        if a[i] > m:
            m = a[i]

    return m

print(fun({"a": 5, "b": 12, "c": 7}))
```
