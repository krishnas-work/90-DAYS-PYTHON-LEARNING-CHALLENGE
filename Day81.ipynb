```python
# Q1: Check if the number of digits is greater than the number of vowels.

# sol:
a = "i have a 4 year old dog but his birth date is 8 may"

d = 0
v = 0

for i in a:
    if i.isdigit():
        d = d + 1
    elif i in "aeiou":
        v = v + 1

if d > v:
    print("digits more")
else:
    print("vowels more or equal")
```

```python
# Q2: Create a new list containing elements whose index is even and value is greater than 5.

# Sol:
a = [3, 7, 6, 2, 9, 4]
b = []

for i in range(len(a)):
    if i % 2 == 0:
        if a[i] > 5:
            b.append(a[i])

print(b)
```

```python
# Q3: Count how many elements are greater than the previous element.

# Sol:
a = (2, 5, 3, 8, 6)
c = 0

for i in range(1, len(a)):
    if a[i] > a[i - 1]:
        c = c + 1

print(c)
```

```python
# Q4: Create a new set containing elements whose square is less than 50.

# sol:
a = {3, 5, 8, 2, 7}
b = set()

for i in a:
    if i * i < 50:
        b.add(i)

print(b)
```

```python
# Q5: Create a function that takes a list and returns the count of numbers whose last digit is even.

# Sol:
def fun(a):
    c = 0

    for i in a:
        if (i % 10) % 2 == 0:
            c = c + 1

    return c

print(fun([12, 35, 47, 28, 50]))
```
