```python
# Q1: Count how many characters in a string are consonants OR digits less than 4.

# sol:
a = "I am 19 years old"
c = 0

for i in a:
    if i.isalpha():
        if i.lower() not in "aeiou":
            c = c + 1
    elif i.isdigit():
        if int(i) < 4:
            c = c + 1

print(c)
```

```python
# Q2: Create a new list containing numbers that are greater than 5 and not equal to 10.


# sol:
a = [4, 6, 10, 8, 3, 12]
b = []

for i in a:
    if i > 5:
        if i != 10:
            b.append(i)

print(b)
```

```python
# Q3: Count how many elements are less than 8 and odd.

# sol:
a = (3, 9, 5, 2, 7, 10)
c = 0

for i in a:
    if i < 8:
        if i % 2 != 0:
            c = c + 1

print(c)
```

```python
# Q4: Create a new set containing elements whose value × 2 is greater than 8.

# sol:
a = {2, 3, 4, 5, 6}
b = set()

for i in a:
    if i * 2 > 8:
        b.add(i)

print(b)
```

```python
# Q5: Create a function that takes a dictionary and returns the sum of values less than 10.

# sol:
def fun(a):
    s = 0

    for i in a:
        if a[i] < 10:
            s = s + a[i]

    return s

print(fun({"a": 5, "b": 12, "c": 7, "d": 15}))
```
