```python
# Q2: Create a list of numbers and:
#     print the maximum and minimum
#     check whether the list length is greater than 5 AND less than 10

# sol:
q=["12", "23", "34", "45", "56", "67", "78"]
if len(q)<10 and len(q)>5:
    print("The length of the list is greater than 5 and less than 10")
else:
    print("The length of the list is not greater than 5 and less than 10")
```

```python
# Q3: Create a tuple with duplicate values and:
#     convert it into a list
#     convert the list into a set
#     print the number of unique elements

# sol:
s=(11,22,33,44,55,66,22,33)
w=list(s)
print("tuple to list:", w)
r=set(w)
print("unique elements:", r)
```

```python
# Q4: Create a dictionary of items and prices and:
#     print all keys , print all values , check whether a given key exists or not

# sol: 
d= {1 : "IS" , 2 : "FOR" , 3: "ME" , "age" : "22"}
q=d.keys()
print(q)
w=d.values()
print(w)

y=input("Enter a key:")
if y in q:
    print("Yes it is exists")
else:
    print("NO it donot exists")
```

```python
# Q5:  Create a list of strings, a set of allowed strings
#      Check whether any element of the list exists in the set and print "Allowed" or "Not Allowed".

# sol:
list1 = ["cat", "dog", "fish"]
set1 = {"lion", "dog", "tiger"}

for item in list1:
    if item in set1:
        print("Allowed")
else:
    print("Not Allowed")
```
