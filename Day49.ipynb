```python
# Q1: Count how many characters in a string are letters but not vowels (i.e., consonants only, ignore non-letters).

# sol:
a = "PyTh0n Programming!"
c = 0

for i in a:
    if i.isalpha() and i.lower() not in "aeiou":
        c = c + 1

print(c)
```

```python
# Q2: Create a new list containing numbers whose sum with the next element is greater than 20.

# sol:
a = [5, 10, 15, 3, 18]
b = []

for i in range(len(a) - 1):
    if a[i] + a[i + 1] > 20:
        b.append(a[i])

print(b)
```

```python
# Q3: Count how many elements in a tuple have more than one digit.

# sol:
a = (5, 12, 7, 105, 9, 33)
c = 0

for i in a:
    if i > 9 or i < -9:
        c = c + 1

print(c)
```

```python
# Q4: Create a new set containing elements whose square is less than 50.

# sol:
a = {3, 5, 7, 8}
b = set()

for i in a:
    if i * i < 50:
        b.add(i)

print(b)
```

```python
# Q5: Create a function that takes a list and returns the sum of numbers whose last digit is even.

# sol:
def total(a):
    s = 0

    for i in a:
        if (i % 10) % 2 == 0:
            s = s + i

    return s

print(total([12, 25, 30, 47, 58]))
```
