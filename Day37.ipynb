```python
# Q1: Count how many spaces are present in a string.

# sol:
a = "python is very easy to learn"
c = 0

for i in a:
    if i == " ":
        c = c + 1

print(c)
```

```python
# Q2: Create a new list containing square of only odd numbers from the list.

# sol:
a = [2, 3, 4, 5, 6, 7]
b = []

for i in a:
    if i % 2 != 0:
        b.append(i * i)

print(b)
```

```python
# Q3: Find the largest element in a tuple using a loop.

# sol:
a = (10, 45, 23, 67, 12)

m = a[0]

for i in a:
    if i > m:
        m = i

print(m)
```

```python
# Q4: Print the keys whose values are greater than 20.

# sol:
a = {"x": 10, "y": 25, "z": 40, "p": 15}

for i in a:
    if a[i] > 20:
        print(i)
```

```python
# Q5: Create a function that takes a list andreturns the sum of all odd numbers

# sol:
def total(a):
    s = 0
    for i in a:
        if i % 2 != 0:
            s = s + i
    return s

print(total([2, 3, 5, 8, 7]))
```
