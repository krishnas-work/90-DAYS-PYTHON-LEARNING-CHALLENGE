```python
# Q1: Find how many characters in a string are letters (alphabets).

# sol:
a = "pyth0n pr0gram"
c = 0

for i in a:
    if i.isalpha():
        c = c + 1

print(c)
```

```python
# Q2: From a list, create a new list containing numbers that are greater than the previous element.

# sol:
a = [3, 5, 2, 8, 7]
b = []

for i in range(1, len(a)):
    if a[i] > a[i-1]:
        b.append(a[i])

print(b)
```

```python
# Q3: Count how many elements in a tuple are between 10 and 30.

# sol:
a = (5, 12, 25, 40, 18, 7)
c = 0

for i in a:
    if i > 10 and i < 30:
        c = c + 1

print(c)
```

```python
# Q4: Create a new set that contains square of each element from the original set.

# sol:
a = {2, 3, 4, 5}
b = set()

for i in a:
    b.add(i * i)

print(b)
```

```python
# Q5: Create a function that takes a list and returns the largest number in the list.

# sol:
def largest(a):
    m = a[0]

    for i in a:
        if i > m:
            m = i

    return m

print(largest([4, 9, 2, 7, 5]))
```
