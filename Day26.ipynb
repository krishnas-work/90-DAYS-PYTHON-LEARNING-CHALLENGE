```python
# Q1: Count how many words are present in a string.

# sol: 
a="i love to learn python"
b=0
for i in a.split():
    b+=1
print(b)
```

```python
# Q2: From a list, create a new list of numbers that are less than the last element.

# sol:
a=[11,22,33,44,55,66,77,88,889,75]
b=[]
for i in a:
    if i < a[-1]:
        b.append(i)
print(b)
```

```python
# Q3: Create a new dictionary containing only the items whose values are odd numbers.

# sol:
a = {"a": 2, "b": 7, "c": 4, "d": 9}
b={}
for key, value in a.items():
    if value % 2 != 0:  
        b[key] = value   

print(b)
```

```python
# Q4: Convert a tuple into a set and:
#     count how many elements are in the set.

# sol:
a = (1, 2, 2, 3, 4, 4, 5)
b = set(a)
c = 0
for i in b:
    c = c + 1

print(c)
```

```python
# Q5: Create a function that:
#     takes a list
#     returns the smallest number from the list.

# sol:
def mini(a):
    m = a[0]
    for i in a:
        if i < m:
            m = i
    return m

print(mini([8, 3, 10, 2, 6]))
```
