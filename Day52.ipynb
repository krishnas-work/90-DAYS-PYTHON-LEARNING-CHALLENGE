```python
# Q1: Count how many characters in a string are either vowels or spaces.

# sol:
a = "Python is Amazing"
c = 0

for i in a:
    if i.lower() in "aeiou" or i == " ":
        c = c + 1

print(c)
```

```python
# Q2: Create a new list containing numbers whose sum of digits is even.

# sol:
a = [23, 45, 18, 10, 37]
b = []

for i in a:
    n = i
    s = 0
    while n > 0:
        s = s + (n % 10)
        n = n // 10

    if s % 2 == 0:
        b.append(i)

print(b)
```

```python
# Q3: Find how many elements in a tuple are less than the average of the tuple.

# sol:
a = (10, 20, 30, 40, 50)

s = 0
for i in a:
    s = s + i

avg = s / len(a)

c = 0
for i in a:
    if i < avg:
        c = c + 1

print(c)
```

```python
# Q4: Create a new dictionary containing items whose value's last digit is odd.

# sol:
a = {"a": 12, "b": 35, "c": 48, "d": 57}
b = {}

for i in a:
    if (a[i] % 10) % 2 != 0:
        b[i] = a[i]

print(b)
```

```python
# Q5: Create a function that takes a set and returns a new set containing elements whose double is divisible by 4.

# sol:
def fun(a):
    b = set()

    for i in a:
        if (i * 2) % 4 == 0:
            b.add(i)

    return b

print(fun({2, 3, 4, 5, 6}))
```
