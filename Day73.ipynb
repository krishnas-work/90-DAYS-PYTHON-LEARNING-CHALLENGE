```python
# Q1: Count how many characters in a string are uppercase vowels OR digits that are even.

# sol:
a = "PyThOn246 AEIOU"
c = 0

for i in a:
    if i in "AEIOU":
        c = c + 1
    elif i.isdigit():
        if int(i) % 2 == 0:
            c = c + 1

print(c)
```

```python
# Q2: Create a new list containing numbers that are greater than 1 and not equal to 4.

# sol:
a = [1, 2, 4, 6, 3, 8]
b = []

for i in a:
    if i > 1:
        if i != 4:
            b.append(i)

print(b)
```

```python
# Q3: Count how many elements are less than 12 and odd.

# sol:
a = (5, 14, 9, 2, 11, 20)
c = 0

for i in a:
    if i < 12:
        if i % 2 != 0:
            c = c + 1

print(c)
```

```python
# Q4: Create a new set containing elements whose value × 3 is greater than 15.

# sol: 
a = {2, 4, 6, 3, 5}
b = set()

for i in a:
    if i * 3 > 15:
        b.add(i)

print(b)
```

```python
# Q5: Create a function that takes a list and returns the count of numbers divisible by 5.

# sol:
def fun(a):
    c = 0

    for i in a:
        if i % 5 == 0:
            c = c + 1

    return c

print(fun([5, 7, 10, 3, 20]))
```
