```python
# Q1: Count how many special characters (not letters or digits) are in a string.

# sol:
a = "Hi@123#ok!"
c = 0

for i in a:
    if not i.isalnum():
        c = c + 1

print(c)
```

```python
# Q2: From a list of numbers:
#     create a new list of numbers that are at even index positions.

# sol:
a = [10, 20, 30, 40, 50, 60]
b = []

for i in range(len(a)):
    if i % 2 == 0:
        b.append(a[i])

print(b)
```

```python
# Q3: Find the difference between the maximum and minimum values in a dictionary.

# sol:
a = {"a": 10, "b": 40, "c": 25}

m1 = a["a"]
m2 = a["a"]

for i in a:
    if a[i] > m1:
        m1 = a[i]
    if a[i] < m2:
        m2 = a[i]

print(m1 - m2)
```

```python
# Q4: From a list:
#     create a set
#     print how many unique elements are present.

# sol:
a = [1, 2, 2, 3, 4, 4, 5]
b = set(a)

c = 0
for i in b:
    c = c + 1

print(c)
```

```python
# Q5: Create a function that:
#     takes a number
#     prints the square of each number from 1 to that number.

# sol:
def show(n):
    for i in range(1, n + 1):
        print(i * i)

show(5)
```
