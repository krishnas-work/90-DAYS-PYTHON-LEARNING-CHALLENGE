```python
# Q1: Count how many digits are present in a string

# sol: 
a = "ab12cd34gs66yjh64rch"
c = 0

for i in a:
    if i.isdigit():
        c = c + 1

print(c)
```

```python
# Q2: From a list, create a new list of numbers greater than the first element.

# Sol:
a = [5, 2, 10, 3, 8]
b = []

for i in a:
    if i > a[0]:
        b.append(i)

print(b)
```

```python
# Q3: Count how many values in a dictionary are even numbers.

# sol:
a = {"a": 2, "b": 5, "c": 8, "d": 7}
c = 0

for i in a:
    if a[i] % 2 == 0:
        c = c + 1

print(c)
```

```python
# Q4: Create a function that:
#     takes a list
#     prints only the even numbers.

# sol:
def show(a):
    for i in a:
        if i % 2 == 0:
            print(i)

show([3, 6, 8, 5, 10])
```

```python
# Q5: Create a function that:
#     takes a list of numbers
#     returns how many elements are greater than 5

# sol: 
def count(a):
    c = 0
    for i in a:
        if i > 5:
            c = c + 1
    return c

print(count([2, 6, 9, 3, 8]))
```
