```python
# Q1: Count how many characters in a string are uppercase letters OR digits equal to 1.

# sol:
a = "IS COding Tough"
c = 0

for i in a:
    if i.isupper():
        c = c + 1
    elif i.isdigit():
        if i == "1":
            c = c + 1

print(c)
```

```python
# Q2: Create a new list containing numbers that are less than 10 and not equal to 6.

# sol:
a = [3, 6, 9, 12, 5, 7]
b = []

for i in a:
    if i < 10:
        if i != 6:
            b.append(i)

print(b)
```

```python
# Q3: Count how many elements are greater than 5 and odd/

# sol:
a = (2, 7, 10, 9, 4, 11)
c = 0

for i in a:
    if i > 5:
        if i % 2 != 0:
            c = c + 1

print(c)
```

```python
# Q4: Create a new set containing elements whose value × 2 is less than 12.

# sol:
a = {3, 4, 5, 6, 7}
b = set()

for i in a:
    if i * 2 < 12:
        b.add(i)

print(b)
```

```python
# Q5: Create a function that takes a list and returns the count of numbers divisible by 2.

# Sol:
def fun(a):
    c = 0

    for i in a:
        if i % 2 == 0:
            c = c + 1

    return c

print(fun([2, 5, 8, 7, 6]))
```
