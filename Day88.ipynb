```python
# Q1: Check if number of digits divisible by 2 + uppercase vowels is less than number of lowercase consonants.

# sol:
a = "PyTHon is a Sanke3213"

x = 0
y = 0

for i in a:
    if i.isdigit():
        if int(i) % 2 == 0:
            x = x + 1
    elif i.isalpha():
        if i.isupper():
            if i in "AEIOU":
                x = x + 1
    elif i.isalpha():
        if i.islower():
            if i not in "aeiou":
                y = y + 1

if x < y:
    print("yes")
else:
    print("no")
```

```python
# Q2: Create a new list of elements where value is greater than (index + 2) and less than (index × 3 + 5).

# sol:
a = [3, 6, 9, 12, 5, 15]
b = []

for i in range(len(a)):
    if a[i] > i + 2:
        if a[i] < i * 3 + 5:
            b.append(a[i])

print(b)
```

```python
# Q3: Count elements where difference with previous is even and difference with next is odd.

# Sol:
a = (4, 7, 5, 10, 9)
c = 0

for i in range(1, len(a) - 1):
    diff1 = a[i] - a[i - 1]
    diff2 = a[i + 1] - a[i]

    if diff1 % 2 == 0 and diff2 % 2 != 0:
        c = c + 1

print(c)
```

```python
# Q4: Create a new set of elements whose (value² + 2×value) is divisible by 4 and less than 40.

# sOL:
a = {2, 3, 4, 5, 6}
b = set()

for i in a:
    val = i * i + 2 * i

    if val % 4 == 0:
        if val < 40:
            b.add(i)

print(b)
```

```python
# Q5: Create a function that takes a list and returns numbers whose sum of digits is greater than first digit.

# Sol:
def fun(a):
    b = []

    for i in a:
        temp = i
        s = 0

        while temp > 0:
            s = s + (temp % 10)
            temp = temp // 10

        first = i
        while first >= 10:
            first = first // 10

        if s > first:
            b.append(i)

    return b

print(fun([23, 41, 56, 72, 19]))
```
