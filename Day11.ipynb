```python
# Q1: Take a string and check whether it contains more vowels than consonants.

# sol:
a = "his name is vaurn thakur and he is a very good boy."

vowels = 0
consonants = 0

for i in a.lower():
    if i.isalpha():  
        if i in "aeiou":
            vowels = vowels+1
        else:
            consonants = consonants+1

print("Vowels:", vowels)
print("Consonants:", consonants)

if vowels > consonants:
    print("The string has more vowels than consonants")
else:
    print("The string does NOT have more vowels than consonants")
```

```python
# Q2: Create a list of numbers and:
#     count how many numbers are positive and even 
#     count how many numbers are negative or odd

# sol:

a =[1, -2, 3, 4, -5, 6, -7, 8, 0]
m= 0
n= 0  
for x in a:
    if x > 0 and x % 2 == 0:
        m=m+1
    if x < 0 or x % 2 == 1:
         n=n+1

print("List:", a)
print("Positive even:", m)
print("Negative or odd:", n)
    
```

```python
# Q4: Take a string as input and:
#     convert the string into a list of characters
#     convert the list into a set
#     print all unique characters using a for loop 

# sol:
a = input("Enter a string: ")

b = []      
c = set()   


for i in a:
    b.append(i)
for j in b:
    c.add(j)
    
print("Unique characters:")
for k in c:
    print(k)


```

```python
# Q5: Create a list of words and:
#     store each word in a dictionary
#     store the count of each word as its value

# sol: 
words = ["cat", "dog", "cat", "cow", "dog", "cat"]
d = {} 
for w in words:
    if w in d:
        d[w] = d[w] + 1
    else:
        d[w] = 1
print(d)
```
