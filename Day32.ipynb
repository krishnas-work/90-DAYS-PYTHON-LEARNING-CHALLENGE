```python
# Q1: Count how many words have more than 3 letters in a string.

# sol:
a = "I love learning python coding"
c = 0

for i in a.split():
    if len(i) > 3:
        c = c + 1

print(c)
```

```python
# Q2: From a list of numbers:
#     create a new list of numbers that are less than the middle element  , (assume list has odd number of elements)

# sol:
a = [3, 8, 5, 10, 7]
mid = a[len(a)//2]

b = []
for i in a:
    if i < mid:
        b.append(i)

print(b)
```

```python
# Q3: Find how many values in a dictionary are perfect squares.

# sol:
a = {"x": 4, "y": 6, "z": 9, "w": 10}
c = 0

for i in a:
    v = a[i]
    if int(v**0.5) * int(v**0.5) == v:
        c = c + 1

print(c)
```

```python
# Q4: From a tuple, create a list of elements that are between 5 and 15.

# sol:
a = (2, 6, 12, 20, 9, 3)

b = []
for i in a:
    if i > 5 and i < 15:
        b.append(i)

print(b)
```

```python
# Q5: Create a function that:
#     takes a number
#      returns the count of digits using while loop.

# sol:
def count_digits(n):
    c = 0
    while n > 0:
        c = c + 1
        n = n // 10
    return c

print(count_digits(12345))
```
