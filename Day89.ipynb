```python
# Q1: Check if (digits greater than 4 + lowercase vowels) is greater than (uppercase consonants + digits equal to 1).


# sol:
a = " The 90 Days challenge is about to get completed"

x = 0
y = 0

for i in a:
    if i.isdigit():
        if int(i) > 4:
            x = x + 1
        if i == "1":
            y = y + 1
    elif i.isalpha():
        if i in "aeiou":
            x = x + 1
        elif i.isupper() and i not in "AEIOU":
            y = y + 1

if x > y:
    print("yes")
else:
    print("no")
```

```python
# Q2: Create a new list of elements where value is greater than (index × 2 + 1) and less than (index² + 10).

# Sol:
a = [3, 7, 5, 12, 8, 20]
b = []

for i in range(len(a)):
    if a[i] > i * 2 + 1:
        if a[i] < i * i + 10:
            b.append(a[i])

print(b)
```

```python
# Q3: Count elements where greater than previous less than next and even.

# sol:
a = (3, 6, 5, 10, 7, 12)
c = 0

for i in range(1, len(a) - 1):
    if a[i] > a[i - 1] and a[i] < a[i + 1]:
        if a[i] % 2 == 0:
            c = c + 1

print(c)
```

```python
# Q4: Create a new set of elements whose (value³ − value) is divisible by 5 and greater than 20.

# sol:
a = {2, 3, 4, 5, 6}
b = set()

for i in a:
    val = i * i * i - i

    if val % 5 == 0:
        if val > 20:
            b.add(i)

print(b)
```

```python
# Q5: Create a function that takes a list and returns numbers where sum of first half digits > sum of second half digits.

# sol:
def fun(a):
    b = []

    for i in a:
        digits = []
        temp = i

        while temp > 0:
            digits.append(temp % 10)
            temp = temp // 10

        digits.reverse()

        mid = len(digits) // 2

        s1 = 0
        s2 = 0

        for j in range(mid):
            s1 = s1 + digits[j]

        for j in range(mid, len(digits)):
            s2 = s2 + digits[j]

        if s1 > s2:
            b.append(i)

    return b

print(fun([1234, 5612, 9870, 4523]))
```
