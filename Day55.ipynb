```python
# Q1: Count how many characters in a string are letters but not lowercase vowels.

# sol:
a = "PyThOn PrOgram"
c = 0

for i in a:
    if i.isalpha() and i.lower() not in "aeiou":
        c = c + 1

print(c)
```

```python
# Q2: Create a new list containing numbers whose difference between digits is greater than 2.

# sol:
a = [23, 45, 19, 82, 56]
b = []

for i in a:
    x = i % 10
    y = i
    while y >= 10:
        y = y // 10

    if abs(x - y) > 2:
        b.append(i)

print(b)
```

```python
# Q3: Count how many elements in a tuple are greater than the average and even.

# sol:
a = (10, 15, 20, 25, 30)

s = 0
for i in a:
    s = s + i

avg = s / len(a)

c = 0
for i in a:
    if i > avg and i % 2 == 0:
        c = c + 1

print(c)
```

```python
# Q4: Create a new set containing elements whose cube minus the number is divisible by 5.

# sol:
a = {2, 3, 4, 5}
b = set()

for i in a:
    if (i**3 - i) % 5 == 0:
        b.add(i)

print(b)
```

```python
# Q5: Create a function that takes a dictionary and returns the sum of values whose first digit is even.

# sol:
def total(a):
    s = 0

    for i in a:
        n = a[i]

        while n >= 10:
            n = n // 10

        if n % 2 == 0:
            s = s + a[i]

    return s

print(total({"a": 23, "b": 48, "c": 62, "d": 15}))
```
