```python
# Q1: Count how many words end with a vowel in a string.

# sol:
a = "banana apple dog orange"
c = 0

for i in a.split():
    if i[-1] in "aeiou":
        c = c + 1

print(c)
```

```python
# Q2: From a list, create a new list that contains only unique elements

# sol:
a = [2, 3, 2, 5, 3, 7]
b = []

for i in a:
    if i not in b:
        b.append(i)

print(b)
```

```python
# Q3: Create a new dictionary by swapping keys and values

# sol:
a = {"x": 10, "y": 20, "z": 30}
b = {}

for i in a:
    b[a[i]] = i

print(b)
```

```python
# Q4: From a tuple, count how many elements are divisible by 4.

# sol:
a = (4, 7, 8, 12, 5)
c = 0

for i in a:
    if i % 4 == 0:
        c = c + 1

print(c)
```

```python
# Q5: Create a function that ,takes a list and returns a new list with double of each element.

# sol:
def double(a):
    b = []
    for i in a:
        b.append(i * 2)
    return b

print(double([2, 4, 6]))
```
