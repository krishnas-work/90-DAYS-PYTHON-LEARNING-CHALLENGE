```python
# Q1: Find how many characters in a string are digits.

# sol:
a = "abc123de45"
c = 0

for i in a:
    if i.isdigit():
        c = c + 1

print(c)
```

```python
# Q2: Create a list that contains only numbers greater than the average of the list.

# sol:
a = [2, 4, 6, 8]
s = 0

for i in a:
    s = s + i

avg = s / len(a)
b = []

for i in a:
    if i > avg:
        b.append(i)

print(b)
```

```python
# Q3: Count how many values in a dictionary are even numbers.

# sol: 
a = {"a": 2, "b": 5, "c": 8, "d": 7}
c = 0

for i in a:
    if a[i] % 2 == 0:
        c = c + 1

print(c)
```

```python
# Q4: Find common elements between two sets using loop.

# sol:
a = {2, 4, 6, 8}
b = {4, 8, 10}
c = set()

for i in a:
    if i in b:
        c.add(i)

print(c)
```

```python
# Q5: Create a function that takes a string and returns number of uppercase letters.

# sol:
def upper_count(a):
    c = 0
    for i in a:
        if i.isupper():
            c = c + 1
    return c

print(upper_count("PyTHon"))
```
