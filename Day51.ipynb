```python
# Q1: Count how many characters in a string are either consonants or digits.

# sol:
a = "PyTh0n 3.0.12 is Fun"
c = 0

for i in a:
    if (i.isalpha() and i.lower() not in "aeiou") or i.isdigit():
        c = c + 1

print(c)
```

```python
# Q2: Create a new list containing numbers whose product with the next element is greater than 50.

# sol:
a = [2, 5, 10, 3, 20]
b = []

for i in range(len(a) - 1):
    if a[i] * a[i + 1] > 50:
        b.append(a[i])

print(b)
```

```python
# Q3: Count how many elements in a tuple have exactly 2 digits.

# sol:
a = (5, 12, 123, 45, 9, 88)
c = 0

for i in a:
    if (i >= 10 and i <= 99) or (i <= -10 and i >= -99):
        c = c + 1

print(c)
```

```python
# Q4: Create a new set containing elements whose square + 1 is greater than 50.

# sol:
a = {5, 6, 7, 8}
b = set()

for i in a:
    if i*i + 1 > 50:
        b.add(i)

print(b)
```

```python
# Q5: Create a function that takes a list and returns the sum of numbers whose first digit is odd.

# sol:
def total(a):
    s = 0

    for i in a:
        x = i
        while x >= 10:
            x = x // 10

        if x % 2 != 0:
            s = s + i

    return s

print(total([23, 46, 51, 78, 95]))
```
