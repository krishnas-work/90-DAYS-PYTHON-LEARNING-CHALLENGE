```python
# Q1: Take a string as input and:
#     use a for loop to count how many characters are digits and how many are letters
#     print both counts

# sol:
o=" their are 15 students present in the class, but total strength of class is 30"
digitcount=0
lettercount=0
for i in o:
    if i.isdigit():
        digitcount=digitcount+1
    elif i.isalpha():
        lettercount=lettercount+1
print("no. of digits:", digitcount )
print("no. of letters:", lettercount )
```

```python
# Q2: Create a list of numbers and:
#     use a while loop to go through the list
#     count how many numbers are greater than 10
#     print the count

# sol:
t=[87,5,43,55,23,1,9,0,71,94,62,3]
i=0
while len(t) > 0:
    e= t.pop()
    if e > 10:
        i=i+1
print("numbers greater than 10 are" , i)
```

```python
# Q3: Create a tuple with duplicate values and:
#     convert it into a set
#     use a for loop to print only the unique elements

#  sol:
f=(11,22,33,44,22,11)
w=[]
for i in f:
    if i not in w:
        print(i)
        w.append(i)
```

```python
# Q4: Create a dictionary of student names and marks and:
#     using a for loop, print the name of students who scored between 40 and 70
#     print "Out of range" for others

a = {"sia": 42, "ria": 33, "pia": 54, "kia": 66, "mia": 69}

for i in a:
    if a[i] > 40 and a[i] < 70:
        print(i)
    else:
        print("Out of range")
```

```python
# Q5: Take a string as input and:
#     convert it into a list of characters
#     convert the list into a set
#     using a while loop, print all the unique characters

# sol:
a = input("Enter a string: ")
l = list(a)
s = set(l)
i = 0
p = list(s)
while i < len(p):
    print(p[i])
    i += 1
```
