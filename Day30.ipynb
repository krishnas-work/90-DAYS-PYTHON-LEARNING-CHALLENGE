```python
# Q1: Count how many words start with a vowel in a string.

# sol: 
a = "apple is on orange table"
c = 0

for i in a.split():
    if i[0] in "aeiou":
        c = c + 1

print(c)
```

```python
# Q2: From a list of numbers:
#     create a new list of numbers that are greater than the sum of first two elements.

# sol:
a = [2, 3, 10, 5, 8]
b = []

s = a[0] + a[1]

for i in a:
    if i > s:
        b.append(i)

print(b)
```

```python
# Q3: Create a new dictionary where:
#     keys remain the same
#     values become double of original values.

# sol:
a = {"x": 5, "y": 8, "z": 3}
b = {}

for i in a:
    b[i] = a[i] * 2

print(b)
```

```python
# Q4: From a tuple:
#     count how many elements are negative numbers.

# sol:
a = (3, -1, 4, -5, 2)

c = 0
for i in a:
    if i < 0:
        c = c + 1

print(c)
```

```python
# Q5: Create a function that:
#     takes a number
#     prints only the even digits of that number using while loop.

# sol: 
def show(n):
    while n > 0:
        d = n % 10
        if d % 2 == 0:
            print(d)
        n = n // 10

show(48237)
```
