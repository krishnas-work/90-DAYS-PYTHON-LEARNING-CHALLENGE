```python
# Q1: Count how many characters in a string are not letters.

# sol:
a = "Python3.10 is cool!"
c = 0

for i in a:
    if not i.isalpha():
        c = c + 1

print(c)
```

```python
# Q2: Create a new list containing numbers whose remainder when divided by 4 is 1.

# sol:
a = [5, 8, 9, 13, 16, 21]
b = []

for i in a:
    if i % 4 == 1:
        b.append(i)

print(b)
```

```python
# Q3: Count how many elements in a tuple are perfect squares.

# sol:
a = (4, 7, 9, 10, 16, 18)
c = 0

for i in a:
    r = int(i ** 0.5)
    if r * r == i:
        c = c + 1

print(c)
```

```python
# Q4: Create a new set containing elements whose value minus 2 is greater than 5

# sol:
a = {4, 6, 8, 10, 12}
b = set()

for i in a:
    if i - 2 > 5:
        b.add(i)

print(b)
```

```python
# Q5: Create a function that takes a dictionary and returns the number of values greater than 50.

# sol:
def count_val(a):
    c = 0

    for i in a:
        if a[i] > 50:
            c = c + 1

    return c

print(count_val({"a":40, "b":55, "c":70, "d":20}))
```
