```python
# Q1: Count how many characters in a string are lowercase letters OR digits that are odd.

# sol:
a = "PyThon135 Code"
c = 0

for i in a:
    if i.islower():
        c = c + 1
    elif i.isdigit():
        if int(i) % 2 != 0:
            c = c + 1

print(c)
```

```python
# Q2: Create a new list containing numbers that are greater than 2 and not equal to 9

# sol:
a = [1, 3, 9, 5, 2, 7]
b = []

for i in a:
    if i > 2:
        if i != 9:
            b.append(i)

print(b)
```

```python
# Q3: Count how many elements are greater than 4 and even

# sol:
a = (2, 6, 5, 8, 3, 10)
c = 0

for i in a:
    if i > 4:
        if i % 2 == 0:
            c = c + 1

print(c)
```

```python
# Q4: Create a new set containing elements whose value + 1 is less than 6

# sol:
a = {2, 3, 5, 6, 1}
b = set()

for i in a:
    if i + 1 < 6:
        b.add(i)

print(b)
```

```python
# Q5: Create a function that takes a list and returns the sum of numbers divisible by 2.

# sol:
def fun(a):
    s = 0

    for i in a:
        if i % 2 == 0:
            s = s + i

    return s

print(fun([2, 5, 6, 7, 8]))
```
