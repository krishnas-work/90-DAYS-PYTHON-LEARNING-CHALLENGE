```python
# Q1: Count how many characters in a string are uppercase letters OR digits less than 5.

# sol:
a = "PyTHon243 CODE7"
c = 0

for i in a:
    if i.isupper():
        c = c + 1
    elif i.isdigit():
        if int(i) < 5:
            c = c + 1

print(c)
```

```python
# Q2: Create a new list containing numbers that are greater than 3 and not equal to 8.

# sol:
a = [2, 4, 8, 5, 1, 9]
b = []

for i in a:
    if i > 3:
        if i != 8:
            b.append(i)

print(b)
```

```python
# Q3: Count how many elements are greater than 2 and less than 10.

# sol:
a = (1, 5, 12, 3, 9, 2)
c = 0

for i in a:
    if i > 2:
        if i < 10:
            c = c + 1

print(c)
```

```python
# Q4: Create a new set containing elements whose value − 1 is greater than 3,

# sol:
a = {2, 4, 6, 3, 5}
b = set()

for i in a:
    if i - 1 > 3:
        b.add(i)

print(b)
```

```python
# Q5: Create a function that takes a dictionary and returns the count of values divisible by 3.

# sol:
def fun(a):
    c = 0

    for i in a:
        if a[i] % 3 == 0:
            c = c + 1

    return c

print(fun({"a": 3, "b": 7, "c": 9, "d": 5}))
```
