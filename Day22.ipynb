```python
# Q1: Count how many spaces are in a string.

# sol:
a = "I love python"
c = 0

for i in a:
    if i == " ":
        c = c + 1

print(c)
```

```python
# Q2: From a list, create a new list of numbers less than 10.

# sol:
a = [5, 12, 3, 20, 7]
b = []

for i in a:
    if i < 10:
        b.append(i)

print(b)
```

```python
# Q3: Print the names of students who passed (marks â‰¥ 40).

# sol:
a = {"Ram": 35, "Sam": 60, "Tom": 45}

for i in a:
    if a[i] >= 40:
        print(i)
```

```python
# Q4: Create a function that prints numbers from 1 to 5 using while loop.

#sol:
def show():
    i = 1
    while i <= 5:
        print(i)
        i = i + 1

show()
```

```python
# Q5: Take a string and:
#     count how many special characters are present

# sol:
a = "hello@1234567*()!"
c = 0

for i in a:
    if not i.isalnum():
        c = c + 1

print(c)
```
