```python
# Q1: Count how many words end with a consonant in a string.

# sol:
a = "apple bat cat orange dog"
c = 0

for i in a.split():
    if i[-1] not in "aeiou":
        c = c + 1

print(c)
```

```python
# Q2: From a list, create a new list by rotating elements to the left by 1.

# sol:
a = [1, 2, 3, 4]
b = []

for i in range(1, len(a)):
    b.append(a[i])

b.append(a[0])
print(b)
```

```python
# Q3: Count how many keys start with a vowel in a dictionary.

# sol:
a = {"apple": 10, "ball": 5, "orange": 8, "cat": 3}
c = 0

for i in a:
    if i[0] in "aeiou":
        c = c + 1

print(c)
```

```python
# Q4: From a tuple:
#     convert it into a set
#     create a list of elements greater than 10

# sol: 
a = (5, 12, 7, 20, 12)
b = set(a)

c = []
for i in b:
    if i > 10:
        c.append(i)

print(c)
```

```python
# Q5: Create a function that:
#     takes a list
#     returns the product of all elements

# sol:
def prod(a):
    p = 1
    for i in a:
        p = p * i
    return p

print(prod([2, 3, 4]))
```
