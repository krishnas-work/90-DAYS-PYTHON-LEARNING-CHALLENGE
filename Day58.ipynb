```python
# Q1: Count how many characters in a string are lowercase letters OR digits less than 5

# sol:
a = "PyThon324 Code9"
c = 0

for i in a:
    if i.islower():
        c = c + 1
    elif i.isdigit():
        if int(i) < 5:
            c = c + 1

print(c)
```

```python
# Q2: Create a new list containing numbers whose last digit is greater than first digit

# sol:
a = [23, 41, 56, 72, 19]
b = []

for i in a:
    last = i % 10

    n = i
    while n >= 10:
        n = n // 10

    first = n

    if last > first:
        b.append(i)

print(b)
```

```python
# Q3: Count how many elements in a tuple are divisible by both 2 and 3

# Sol:
a = (6, 10, 12, 15, 18, 7)
c = 0

for i in a:
    if i % 2 == 0 and i % 3 == 0:
        c = c + 1

print(c)
```

```python
# Q4: Create a new set containing elements whose value + 3 is even

# sol:
a = {1, 2, 3, 4, 5}
b = set()

for i in a:
    if (i + 3) % 2 == 0:
        b.add(i)

print(b)
```

```python
# Q5: Create a function that takes a dictionary and returns the sum of values that are even and greater than 10.

# Sol:
def fun(a):
    s = 0

    for i in a:
        if a[i] > 10 and a[i] % 2 == 0:
            s = s + a[i]

    return s

print(fun({"a": 12, "b": 7, "c": 20, "d": 9}))
```
