```python
# Q1: Count how many characters in a string are not vowels.

# sol:
a = "programming"
c = 0

for i in a:
    if i not in "aeiou":
        c = c + 1

print(c)
```

```python
# Q2: From a list of numbers, create a new list containing numbers divisible by both 2 and 3.

# sol:
a = [3, 6, 8, 12, 15, 18]
b = []

for i in a:
    if i % 2 == 0 and i % 3 == 0:
        b.append(i)

print(b)
```

```python
# Q3: Count how many numbers in a tuple are less than 10.

# sol:
a = (5, 12, 7, 20, 9, 15)
c = 0

for i in a:
    if i < 10:
        c = c + 1

print(c)
```

```python
# Q4: Create a new set that contains elements from a set greater than 5.

# Sol:
a = {2, 4, 6, 8, 3, 10}
b = set()

for i in a:
    if i > 5:
        b.add(i)

print(b)
```

```python
# Q5: Create a function that takes a string and returns the number of lowercase letters.

# sol:
def lower_count(a):
    c = 0
    for i in a:
        if i.islower():
            c = c + 1
    return c

print(lower_count("PyThOn ProGram"))
```
