```python
# Q1: Count how many characters in a string are consonants.

# sol:
a = "python programming"
c = 0

for i in a:
    if i.isalpha() and i not in "aeiou":
        c = c + 1

print(c)
```

```python
# Q2: Create a new list containing numbers whose last digit is greater than 5.

# sol:
a = [12, 47, 39, 64, 85]
b = []

for i in a:
    if i % 10 > 5:
        b.append(i)

print(b)
```

```python
# Q3: Find the product of all elements in a tuple.

# sol:
a = (2, 3, 4, 5)
p = 1

for i in a:
    p = p * i

print(p)
```

```python
# Q4: Create a list of values that are greater than 50 from a dictionary.

# sol:
a = {"a": 40, "b": 65, "c": 30, "d": 80}
b = []

for i in a:
    if a[i] > 50:
        b.append(a[i])

print(b)
```

```python
# Q5: Create a function that takes a string and returns the string without spaces.

# sol:
def remove_space(a):
    b = ""

    for i in a:
        if i != " ":
            b = b + i

    return b

print(remove_space("python is fun"))
```
