```python
# Q1: Count how many characters in a string are spaces OR uppercase letters

# sol:
a = "Py Thon Is Fun"
c = 0

for i in a:
    if i == " ":
        c = c + 1
    elif i.isupper():
        c = c + 1

print(c)
```

```python
# Q2: Create a new list containing numbers that are greater than 10 and even

# sol:
a = [5, 12, 7, 18, 3, 20]
b = []

for i in a:
    if i > 10:
        if i % 2 == 0:
            b.append(i)

print(b)
```

```python
# Q3: Count how many elements in a tuple are less than 15

# Sol:
a = (10, 25, 5, 18, 12)
c = 0

for i in a:
    if i < 15:
        c = c + 1

print(c)
```

```python
# Q4: Create a new set containing elements that are odd numbers

# sol:
a = {2, 3, 4, 5, 6, 7}
b = set()

for i in a:
    if i % 2 != 0:
        b.add(i)

print(b)
```

```python
# Q5: Create a function that takes a dictionary and returns the count of values greater than 10.

# Sol:
def fun(a):
    c = 0

    for i in a:
        if a[i] > 10:
            c = c + 1

    return c

print(fun({"a": 5, "b": 15, "c": 8, "d": 20}))
```
