```python
# Q1: Write a function that:
#     takes a number
#     prints whether it is even or odd

# sol: 
def check(x):
    if x % 2 == 0:
        print("Even")
    else:
        print("Odd")

y = int(input())
check(y)
```

```python
# Q2:Write a function that:
#    takes a list of numbers
#    counts how many numbers are greater than 10
#     prints the count

# sol: 
def count(x):
    c = 0
    for y in x:
        if y > 10:
            c = c + 1
    print(c)

z = [5, 12, 8, 20, 3]
count(z)
```

```python
# Q3: Take a string and:
#     count how many vowels are present in it

# sol:
x = input()
c = 0

for y in x:
    if y in "aeiouAEIOU":
        c = c + 1

print(c)

```

```python
# Q4: Create a tuple with duplicate values and:
#     convert it into a set
#     use a for loop to print all unique elements

# sol:
x = (1, 2, 2, 3, 4, 4)
y = set(x)

for z in y:
    print(z)
```

```python
# Q5: Create a dictionary of names and marks and:
#     print the names of students who scored 50 or more

# sol:
x = {"Aman": 45, "Riya": 78, "Neha": 50, "Rahul": 30}

for y in x:
    if x[y] >= 50:
        print(y)
```
