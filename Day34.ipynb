```python
# Q1: Find how many words start with a vowel in a string.

# sol: 
a = "apple is on the orange table"
c = 0

for i in a.split():
    if i[0] in "aeiou":
        c = c + 1

print(c)
```

```python
# Q2: From a list of numbers create a new list of numbers whose sum of digits is greater than 5.

#sol:
a = [12, 34, 5, 20, 9]
b = []

for i in a:
    n = i
    s = 0
    while n > 0:
        s = s + (n % 10)
        n = n // 10
    if s > 5:
        b.append(i)

print(b)
```

```python
# Q3: From a dictionary of marks , print the names whose marks are divisible by 5.

# sol:
a = {"ram": 45, "sam": 32, "tom": 50}
for i in a:
    if a[i] % 5 == 0:
        print(i)
```

```python
# Q4: Create a list of elements from a tuple that are greater than the average.

# sol:
a = (10, 20, 30, 40)

s = 0
for i in a:
    s = s + i

avg = s / len(a)

b = []
for i in a:
    if i > avg:
        b.append(i)

print(b)
```

```python
# Q5: Create a function that ,takes a string and returns the number of uppercase letters.

# sol:
def upper_count(a):
    c = 0
    for i in a:
        if i.isupper():
            c = c + 1
    return c

print(upper_count("PyTHon"))
```
