```python
# Q1: Take a string as input and:
#     check whether the string length is greater than 4 AND it contains the letter 'a'
#     if yes, print the string in uppercase
#     otherwise, print the string in lowercase

# sol: 
a=input("Enter a string:")
if len(a)>4 and 'a' in a:
    q="yes"
    print(q)
else:
    w="no"
    print(w)

if q:
    print(a.upper())
else:
    print(a.lower())
        
```

```python
# Q2: Create a list of numbers and use a for loop to:
#      print each number
#      print "Even" if the number is divisible by 2
#      otherwise print "Odd"

# sol:
u= [11,22,33,44,55,66,77,88]
for i in u:
    print(i, end = "  ")
    if i%2==0:
        print("Even")
    else:
        print("odd")
        
```

```python
# Q3: Create a tuple containing duplicate numbers and:
#     convert it into a set
#     use a for loop to print each unique element

# sol:
w=(11,22,33,44,55,66,22,33)
r=set(w)
print(r)
for i in r:
    print(i)
```

```python
# Q4: Create a dictionary of student names and marks and:
#     using a for loop, print the name of students who scored 50 or more
#     print "Fail" for students who scored less than 50

# sol:
d= {"krishna" : 23 , "varun" : 56 , "aditya" : 12 , "yashi" : 67}
for name, marks in d.items():
    if  marks>=50:
        print(f"{name}:{marks}")
    else:
        print(f"{name}: Fail")
```

```python
# Q5: Create:
#     a list of strings
#     a set of allowed strings
#     Using a basic for loop, check:
#     if any string from the list exists in the set, print "Match Found"
#     otherwise print "No Match"

# Sol:
q=["my" , "name" , "is" , "don"]
s={'hello' , 'everyone' , 'this' , 'is' , 'box'}
print(s)
for i in q:
    if i in s:
        print("Match  Found:", i)
    else:
        print("Not Found:", i)
```
