```python
# Q1: Count how many characters in a string are not digits.

# sol:
a = "Py7th0n2026"
c = 0

for i in a:
    if not i.isdigit():
        c = c + 1

print(c)
```

```python
# Q2: Create a new list containing numbers that are multiples of 5.

# Sol:
a = [3, 10, 14, 25, 7]
b = []

for i in a:
    if i % 5 == 0:
        b.append(i)

print(b)
```

```python
# Q3: Find the sum of elements greater than 5.

# sol:
a = (2, 8, 3, 10, 5)
s = 0

for i in a:
    if i > 5:
        s = s + i

print(s)
```

```python
# Q4: Find the sum of elements greater than 5.

# sol:
a = {3, 4, 6, 7, 9, 10}
b = set()

for i in a:
    if i % 3 != 0:
        b.add(i)

print(b)
```

```python
# Q5:Create a function that takes a list and returns the count of numbers greater than 7.

# sol:
def fun(a):
    c = 0

    for i in a:
        if i > 7:
            c = c + 1

    return c

print(fun([5, 8, 3, 10, 6]))
```
