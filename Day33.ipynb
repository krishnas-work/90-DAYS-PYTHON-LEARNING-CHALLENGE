```python
# Q1: Count how many words contain the letter 'e' in a string.

# sol:
a = "we are learning python here"
c = 0

for i in a.split():
    if "e" in i:
        c = c + 1

print(c)
```

```python
# Q2: From a list of numbers create a new list of numbers that are greater than both first and last elements.

# sol:
a = [5, 2, 10, 8, 3]
first = a[0]
last = a[-1]

b = []
for i in a:
    if i > first and i > last:
        b.append(i)

print(b)
```

```python
# Q3: Create a new dictionary where: keys stay the same , values become the sum of digits of the original values.

# sol:
a = {"x": 23, "y": 45, "z": 8}
b = {}

for i in a:
    v = a[i]
    s = 0
    while v > 0:
        s = s + (v % 10)
        v = v // 10
    b[i] = s

print(b)
```

```python
# Q4: From a tuple create a list of elements whose index is even.

# sol:
a = (10, 20, 30, 40, 50)

b = []
for i in range(len(a)):
    if i % 2 == 0:
        b.append(a[i])

print(b)
```

```python
# Q5: Create a function that takes a list , returns the average of all elements.

# sol:
def avg(a):
    s = 0
    for i in a:
        s = s + i
    return s / len(a)

print(avg([10, 20, 30, 40]))
```
