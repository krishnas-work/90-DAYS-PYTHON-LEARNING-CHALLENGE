```python
# Q1: Find the shortest word in a string.

# sol:
a = "python makes coding easier"
short = a.split()[0]

for i in a.split():
    if len(i) < len(short):
        short = i

print(short)
```

```python
# Q2: From a list, create a new list containing numbers whose cube is greater than 100.

# sol:
a = [2, 3, 4, 5]
b = []

for i in a:
    if i**3 > 100:
        b.append(i)

print(b)
```

```python
# Q3: Count how many elements in a tuple are multiples of 5.

# sol:
a = (5, 12, 20, 7, 15, 3)
c = 0

for i in a:
    if i % 5 == 0:
        c = c + 1

print(c)
```

```python
# Q4: Create a new set containing half of each element from the original set.

# sol:
a = {10, 20, 30, 40}
b = set()

for i in a:
    b.add(i / 2)

print(b)
```

```python
# Q5: Create a function that takes a list and returns a list containing only the last digit of each number.

# sol:
def last_digits(a):
    b = []
    
    for i in a:
        b.append(i % 10)
        
    return b

print(last_digits([23, 45, 67, 89]))
```
