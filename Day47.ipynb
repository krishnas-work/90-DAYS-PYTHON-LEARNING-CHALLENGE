```python
# Q1: Count how many characters in a string are either uppercase letters or spaces.

# sol:
a = "Python Is Great"
c = 0

for i in a:
    if i.isupper() or i == " ":
        c = c + 1

print(c)
```

```python
# Q2: Create a new list containing numbers whose half is greater than 6.

# sol:
a = [8, 12, 15, 5, 20]
b = []

for i in a:
    if i / 2 > 6:
        b.append(i)

print(b)
```

```python
# Q3: Find the difference between the largest and smallest element in a tuple.

# sol:
a = (14, 7, 22, 5, 18)

mx = a[0]
mn = a[0]

for i in a:
    if i > mx:
        mx = i
    if i < mn:
        mn = i

print(mx - mn)
```

```python
# Q4: Create a new dictionary containing items whose values are greater than 30.

# sol:
a = {"a": 25, "b": 40, "c": 15, "d": 60}
b = {}

for i in a:
    if a[i] > 30:
        b[i] = a[i]

print(b)
```

```python
# Q5: Create a function that takes a string and returns the number of words in the string.

# sol:
def word_count(a):
    c = 0

    for i in a.split():
        c = c + 1

    return c

print(word_count("python is very easy to learn"))
```
