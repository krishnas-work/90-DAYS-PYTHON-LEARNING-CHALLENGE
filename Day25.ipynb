```python
# Q1: Count how many lowercase letters are in a string.

# sol:
a = "PyTHon"
c = 0

for i in a:
    if i.islower():
        c = c + 1

print(c)
```

```python
# Q2: From a list, create a new list of numbers that are multiples of 4.

# sol:
a = [3, 4, 8, 10, 16, 7]
b = []

for i in a:
    if i % 4 == 0:
        b.append(i)

print(b)
```

```python
# Q3: Find how many values in a dictionary are greater than 20.

# sol:
a = {"x": 15, "y": 25, "z": 40, "w": 10}

c = 0
for i in a:
    if a[i] > 20:
        c = c + 1

print(c)
```

```python
# Q4: From a tuple, create a list of numbers that are divisible by 3.

# sol: 
a = (3, 5, 6, 7, 9, 10)

b = []
for i in a:
    if i % 3 == 0:
        b.append(i)

print(b)
```

```python
# Q5: Create a function that:
#     takes a number
#     prints all numbers from that number down to 1 using while loop.

# sol:
def show(n):
    while n > 0:
        print(n)
        n = n - 1

show(5)
```
