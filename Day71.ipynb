```python
# Q1: Count how many characters in a string are either consonants OR digits.

# sol:
a = "PyTh0n C0de"
c = 0

for i in a:
    if i.isalpha():
        if i.lower() not in "aeiou":
            c = c + 1
    elif i.isdigit():
        c = c + 1

print(c)
```

```python
# Q2: Create a new list containing numbers that are greater than 4 and not equal to 7.

# sol:
a = [2, 5, 7, 9, 3, 8]
b = []

for i in a:
    if i > 4:
        if i != 7:
            b.append(i)

print(b)
```

```python
# Q3: Count how many elements are less than 10 and even.

# sol:
a = (12, 4, 7, 8, 3, 10)
c = 0

for i in a:
    if i < 10:
        if i % 2 == 0:
            c = c + 1

print(c)
```

```python
# Q4: Create a new set containing elements whose value × 2 is less than 10.

# sol:
a = {2, 4, 6, 3, 5}
b = set()

for i in a:
    if i * 2 < 10:
        b.add(i)

print(b)
```

```python
# Q5: Create a function that takes a list and returns the sum of numbers divisible by 4.

# sol:
def fun(a):
    s = 0

    for i in a:
        if i % 4 == 0:
            s = s + i

    return s

print(fun([4, 5, 8, 7, 12]))
```
