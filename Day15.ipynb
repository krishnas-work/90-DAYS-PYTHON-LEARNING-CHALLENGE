```python
# Q1: Take a string as input and:
#     count how many characters are uppercase
#     count how many are lowercase

# sol:
a=input("Enter a string:")
q=0
w=0
for i in a:
    if i.isupper():
        q+=1
    elif i.islower():
        w+=1
print("No. of uppercase:", q)
print("No. of lowercase:", w)
```

```python
# Q3: Create a tuple with duplicate values and:
#     convert it into a set
#     use a while loop to print all unique elements 

# sol:
r=(14,345,245,75,3,6,14,3,6)
e=set(r)
u=list(e)
t=0
while t<len(u):
    print(u[t])
    t=t+1
    

```

```python
# Q4: Create a dictionary of names and ages and:
#     print names of people whose age is 18 or above

# sol:
i = {"danny": 12, "seenu": 23, "bunty": 18, "pia": 17}

for name, age in i.items():
    if age >= 18:
        print(name)
```
