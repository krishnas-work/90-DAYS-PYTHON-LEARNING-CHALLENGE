```python
# Q1: Count how many characters in a string are either lowercase letters or spaces.

# sol:
a = "Python is Fun"
c = 0

for i in a:
    if i.islower() or i == " ":
        c = c + 1

print(c)
```

```python
# Q2: Create a new list containing numbers whose difference with 10 is less than 4.

# sol:
a = [6, 8, 11, 15, 9]
b = []

for i in a:
    if abs(i - 10) < 4:
        b.append(i)

print(b)
```

```python
# Q3: Find the smallest element in a tuple using a loop

# sol:
a = (14, 7, 22, 5, 18)

m = a[0]

for i in a:
    if i < m:
        m = i

print(m)
```

```python
# Q4: Create a new set containing elements whose triple is greater than 20.

# sol:
a = {3, 5, 7, 9}
b = set()

for i in a:
    if i * 3 > 20:
        b.add(i)

print(b)
```

```python
# Q5: Create a function that takes a list and returns the count of numbers divisible by 3.

#  sol: 
def count_num(a):
    c = 0

    for i in a:
        if i % 3 == 0:
            c = c + 1

    return c

print(count_num([3, 5, 9, 10, 12]))
```
