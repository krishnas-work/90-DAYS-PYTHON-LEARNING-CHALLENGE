```python
# Q1: Count how many characters in a string are not equal to the first character.

# sol:
a = "apple"
c = 0

f = a[0]

for i in a:
    if i != f:
        c = c + 1

print(c)
```

```python
# Q2: Create a new list containing numbers that are equal to 0 or greater than 5.

# Sol:
a = [0, 3, 6, 2, 8, 1]
b = []

for i in a:
    if i == 0 or i > 5:
        b.append(i)

print(b)
```

```python
# Q3: Count how many elements are less than the last element.

# Sol:
a = (10, 4, 7, 2, 8)
c = 0

l = a[-1]

for i in a:
    if i < l:
        c = c + 1

print(c)
```

```python
# Q4: Create a new set by subtracting 1 from each element.

# Sol:
a = {3, 5, 7, 9}
b = set()

for i in a:
    b.add(i - 1)

print(b)
```

```python
# Q5: Create a function that and takes a dictionary returns the count of values equal to 5.

# sol:
def fun(a):
    c = 0

    for i in a:
        if a[i] == 5:
            c = c + 1

    return c

print(fun({"a": 5, "b": 3, "c": 5, "d": 7}))
```
