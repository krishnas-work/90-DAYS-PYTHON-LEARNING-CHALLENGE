```python
# Q1: Take a string and:
#     count how many uppercase and lowercase letters are present

# sol:
s="This is the start of the Game"
upper_count= 0
lower_count=0
for i in s:
    if i.isupper():
        upper_count=upper_count+1
    if i.islower():
        lower_count=lower_count+1
        
        
print("Uppercase letters:", upper_count)
print("Lowercase letters:", lower_count)
```

```python
# Q2: From a list of numbers:
#     create a new list of numbers divisible by 3

# sol:
e=[11,22,33,44,55,66,77,88,99,1010]
w=[]
for i in e:
    if i%3==0:
        w.append(i)

print(w)
```

```python
# Q3: From a dictionary of marks:
#    print the name of the student who has the highest marks

# Sol:
o={'sam':33, 'tang':41, 'ram':47}
max_marks = 0
topper = ""

for name, marks in o.items():
    if marks > max_marks:
        max_marks = marks
        topper = name

print("Student with highest marks:", topper)
```

```python
# Q4: Given a list and a set:
#     find common elements using a loop

# sol:
lst = [1, 2, 3, 4, 5]
st = {3, 4, 6, 7}

common = []

for i in lst:
    if i in st:
        common.append(i)

print("Common elements:", common)
```

```python
# Q5: Create a function that:
#     takes a number
#     prints its multiplication table using while loop

# sol:
def table(n):
    i = 1
    while i <= 10:
        print(n, "x", i, "=", n * i)
        i += 1

num = int(input("Enter a number: "))
table(num)
```
