```python
# Q1: Write a program to:
#     Take a string
#     Check if its length is greater than 5

# sol:
a = "python"

if len(a) > 5:
    print("Long")
else:
    print("Short")
```

```python
# Q2: Count how many times each character appears in a string
#     (example: "aba" â†’ {'a':2, 'b':1})

# sol: 
a = "aba"
d = {}

for i in a:
    if i in d:
        d[i] = d[i] + 1
    else:
        d[i] = 1

print(d)
```

```python
# Q3: From a list, create a new list of:
#     only odd numbers

# sol:
a = [1, 2, 3, 4, 5, 6]
b = []

for i in a:
    if i % 2 != 0:
        b.append(i)

print(b)
```

```python
# Q4: Find the key with highest value

# sol:
a = {"x": 10, "y": 50, "z": 30}

m = 0
key = ""

for i in a:
    if a[i] > m:
        m = a[i]
        key = i

print(key)
```

```python
# Q5: Remove all elements of list from a set

# sol: 
a = {1, 2, 3, 4}
b = [2, 4]

for i in b:
    if i in a:
        a.remove(i)

print(a)
```
