```python
# Q1: Take a string as input and print whether its length is greater than 5 or not.
# sol: 
x=input("Enter a string")
y=len(x)
print(y)
if y>5:
    print("The length of the string is greater than 5")
else:
    print("The length of the string is less than 5")
```

```python
# Q3: Take a string and check whether its first character is a vowel or consonant.
# sol:
x=input("Enter a string:")
l=("a,e,i,o,u")
if x.startswith(l):
    print("first character is vowel")
else:
    print("first character is consonant")
```
