```python
# Q1: Count how many characters in a string are not equal to the last character.

# sol:
a = "banana"
c = 0

l = a[-1]

for i in a:
    if i != l:
        c = c + 1

print(c)
```

```python
# Q2: Create a new list containing numbers that are equal to 1 or less than 4.

# sol:
a = [1, 5, 2, 7, 3, 8]
b = []

for i in a:
    if i == 1 or i < 4:
        b.append(i)

print(b)
```

```python
# Q3: Count how many elements are greater than the last element.

# sol:
a = (5, 9, 2, 7, 4)
c = 0

l = a[-1]

for i in a:
    if i > l:
        c = c + 1

print(c)
```

```python
# Q4: Create a new set by adding 5 to each element.

# sol:
a = {1, 2, 3, 4}
b = set()

for i in a:
    b.add(i + 5)

print(b)
```

```python
# Q5: Create a function that takes a list and returns the sum of numbers equal to 2.

# sol:
def fun(a):
    s = 0

    for i in a:
        if i == 2:
            s = s + i

    return s

print(fun([2, 5, 2, 3, 2]))
```
