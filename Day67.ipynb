```python
# Q1: Count how many characters in a string are either "i" or "I".

# sol:
a = "India is Interesting"
c = 0

for i in a:
    if i == "i" or i == "I":
        c = c + 1

print(c)
```

```python
# Q2: Create a new list containing numbers that are greater than 3 and less than 10.

# sol:
a = [2, 5, 8, 12, 1, 7]
b = []

for i in a:
    if i > 3:
        if i < 10:
            b.append(i)

print(b)
```

```python
# Q3: Find the sum of all elements that are even.

# sol:
a = (3, 6, 9, 2, 8)
s = 0

for i in a:
    if i % 2 == 0:
        s = s + i

print(s)
```

```python
# Q4: Create a new set containing elements that are less than or equal to 6.

# sol:
a = {4, 7, 2, 9, 6, 1}
b = set()

for i in a:
    if i <= 6:
        b.add(i)

print(b)
```

```python
# Q5: Create a function that takes a list and returns the count of numbers that are odd.

# sol:
def fun(a):
    c = 0

    for i in a:
        if i % 2 != 0:
            c = c + 1

    return c

print(fun([2, 5, 7, 8, 3]))
```
