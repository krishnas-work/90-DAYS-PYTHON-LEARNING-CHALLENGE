```python
# Q1: Count how many characters in a string are "e" or "E".

# sol:
a = "Elephant is here"
c = 0

for i in a:
    if i == "e" or i == "E":
        c = c + 1

print(c)
```

```python
# Q2: Create a new list containing numbers that are greater than 5.

# sol:
a = [2, 6, 3, 9, 1]
b = []

for i in a:
    if i > 5:
        b.append(i)

print(b)
```

```python
# Q3: Count how many elements are even numbers.

# sol:
a = (4, 7, 10, 3, 8)
c = 0

for i in a:
    if i % 2 == 0:
        c = c + 1

print(c)
```

```python
# Q4: Create a new set by multiplying each element by 2.

# sol:
a = {1, 2, 3, 4}
b = set()

for i in a:
    b.add(i * 2)

print(b)
```

```python
# Q5: Create a function that takes a list and returns the sum of numbers less than 10.

# sol:
def fun(a):
    s = 0

    for i in a:
        if i < 10:
            s = s + i

    return s

print(fun([5, 12, 3, 20, 8]))
```
