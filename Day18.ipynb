```python
# Q1: Write a function that:
#     takes a string
#     counts how many characters are uppercase
#     prints the count

# sol: 
def count_upper(x):
    c = 0
    for y in x:
        if y.isupper():
            c = c + 1
    print(c)

z = input()
count_upper(z)
```

```python
# Q2: Create a list of numbers and:
#     use a while loop to print only the even numbers

# sol:
x = [2, 5, 8, 11, 6]
i = 0

while i < len(x):
    if x[i] % 2 == 0:
        print(x[i])
    i = i + 1
```

```python
# Q3: Create a tuple with duplicate values and:
#     convert it into a list
#     then convert it into a set
#     print the set

# sol:
x = (1, 2, 2, 3, 4, 4)
y = list(x)
z = set(y)

print(z)
```

```python
# Q4: Create a dictionary of subjects and marks and:
#     print subject names whose marks are less than 40

# sol:
x = {"Math": 35, "English": 70, "Science": 30}

for y in x:
    if x[y] < 40:
        print(y)
```

```python
# Q5: Write a function that:
#     uses a while loop
#     prints numbers from 1 to 4

# sol: 
def show():
    i = 1
    while i <= 4:
        print(i)
        i = i + 1

show()
```
