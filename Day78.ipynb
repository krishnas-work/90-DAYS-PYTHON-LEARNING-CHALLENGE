```python
# Q1: Count how many characters in a string are lowercase consonants OR digits equal to 0.

# sol:
a = "PyThon203 Is Easy?"
c = 0

for i in a:
    if i.isalpha():
        if i.islower():
            if i not in "aeiou":
                c = c + 1
    elif i.isdigit():
        if i == "0":
            c = c + 1

print(c)
```

```python
# Q2: Create a new list containing numbers that are greater than 6 and not equal to 11.

# sol:
a = [5, 7, 11, 9, 2, 13]
b = []

for i in a:
    if i > 6:
        if i != 11:
            b.append(i)

print(b)
```

```python
# Q3: Count how many elements are less than 10 and even.

# sol:
a = (2, 11, 6, 9, 4, 15)
c = 0

for i in a:
    if i < 10:
        if i % 2 == 0:
            c = c + 1

print(c)
```

```python
# Q4: Create a new set containing elements whose value + 3 is greater than 8.

# sol:
a = {2, 4, 6, 3, 5}
b = set()

for i in a:
    if i + 3 > 8:
        b.add(i)

print(b)
```

```python
# Q5: Create a function that takes a dictionary and returns the count of values greater than 6.

# sol:
def fun(a):
    c = 0

    for i in a:
        if a[i] > 6:
            c = c + 1

    return c

print(fun({"a": 4, "b": 7, "c": 10, "d": 5}))
```
