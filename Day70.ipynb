```python
# Q1: Count how many characters in a string are either vowels OR spaces.

# sol:
a = "Python is Easy"
c = 0

for i in a:
    if i in "aeiouAEIOU":
        c = c + 1
    elif i == " ":
        c = c + 1

print(c)
```

```python
# Q2: Create a new list containing numbers that are less than 8 and not equal to 5.

# sol:
a = [2, 5, 7, 9, 3, 8]
b = []

for i in a:
    if i < 8:
        if i != 5:
            b.append(i)

print(b)
```

```python
# Q3: Count how many elements are greater than 3 and odd.

# sol:
a = (1, 4, 5, 2, 7, 3)
c = 0

for i in a:
    if i > 3:
        if i % 2 != 0:
            c = c + 1

print(c)
```

```python
# Q4: Create a new set containing elements whose value + 2 is greater than 6.

# sol:
a = {1, 2, 3, 4, 5}
b = set()

for i in a:
    if i + 2 > 6:
        b.add(i)

print(b)
```

```python
# Q5: Create a function that takes a dictionary and returns the count of values divisible by 2.

# sol:
def fun(a):
    c = 0

    for i in a:
        if a[i] % 2 == 0:
            c = c + 1

    return c

print(fun({"a": 4, "b": 3, "c": 6, "d": 5}))
```
