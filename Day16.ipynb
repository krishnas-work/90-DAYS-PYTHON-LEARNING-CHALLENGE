```python
# Q2: Create a list of numbers and:
#     use a for loop to print each number
#     print the square of each number

# sol:
i=[2,4,6,8,3,7]
sq=[]
for w in (i):
    print("number in list:", w)
    sq.append(w**2)

print("Numbers list:", i)
print("Squares list:", sq)
    
```
