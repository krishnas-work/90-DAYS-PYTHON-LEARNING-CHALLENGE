```python
# Q1: Count how many characters in a string are not spaces

# sol:
a = "Python is fun"
c = 0

for i in a:
    if i != " ":
        c = c + 1

print(c)
```

```python
# Q2: Create a new list containing numbers that are less than 20.

# sol:
a = [5, 22, 13, 27, 8, 30]
b = []

for i in a:
    if i < 20:
        b.append(i)

print(b)
```

```python
# Q3: Find the sum of all odd numbers in a tuple

# Sol:
a = (2, 5, 7, 8, 3)
s = 0

for i in a:
    if i % 2 != 0:
        s = s + i

print(s)
```

```python
# Q4: Create a new set containing elements that are greater than 3.

# sol:
a = {1, 2, 3, 4, 5}
b = set()

for i in a:
    if i > 3:
        b.add(i)

print(b)
```

```python
# Q5: Create a function that takes a list and returns the count of even numbers.

# sol:
def fun(a):
    c = 0

    for i in a:
        if i % 2 == 0:
            c = c + 1

    return c

print(fun([2, 5, 8, 7, 6]))
```
