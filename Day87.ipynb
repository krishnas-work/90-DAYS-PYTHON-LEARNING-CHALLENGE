```python
# Q1: Check if number of lowercase vowels + digits less than 5 is greater than number of uppercase consonants.

# sol:
a = "Chcking the vowels3464"

x = 0
y = 0

for i in a:
    if i in "aeiou":
        x = x + 1
    elif i.isdigit():
        if int(i) < 5:
            x = x + 1
    elif i.isalpha():
        if i.isupper():
            if i not in "AEIOU":
                y = y + 1

if x > y:
    print("yes")
else:
    print("no")
```

```python
# Q2: Create a new list of elements where value is greater than index + 1 and less than index + 6.


# sol:
a = [2, 5, 7, 10, 4, 12]
b = []

for i in range(len(a)):
    if a[i] > i + 1:
        if a[i] < i + 6:
            b.append(a[i])

print(b)
```

```python
# Q3: Count elements where it is greater than previous and less than next.

# sol:
a = (3, 6, 5, 9, 7)
c = 0

for i in range(1, len(a) - 1):
    if a[i] > a[i - 1] and a[i] < a[i + 1]:
        c = c + 1

print(c)
```

```python
# Q4: Create a new set of elements whose (value² − value) is divisible by 3 and less than 30.

# Sol:
a = {2, 3, 4, 5, 6}
b = set()

for i in a:
    val = i * i - i

    if val % 3 == 0:
        if val < 30:
            b.add(i)

print(b)
```

```python
# Q5: Create a function that takes a list and returns numbers whose sum of digits is greater than last digit.


# Sol:
def fun(a):
    b = []

    for i in a:
        temp = i
        s = 0

        last = i % 10

        while temp > 0:
            s = s + (temp % 10)
            temp = temp // 10

        if s > last:
            b.append(i)

    return b

print(fun([12, 35, 47, 28, 51]))
```
