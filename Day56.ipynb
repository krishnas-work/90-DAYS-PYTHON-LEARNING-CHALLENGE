```python
# Q1: Count how many characters in a string are either lowercase consonants or odd digits.

# sol:
a = "PyThon135 Code"
c = 0

for i in a:
    if (i.isalpha() and i.islower() and i not in "aeiou") or (i.isdigit() and int(i) % 2 != 0):
        c = c + 1

print(c)
```

```python
# Q2: Create a new list containing numbers whose sum of digits is divisible by 3 and greater than 5.

# sol:
a = [12, 23, 45, 18, 29]
b = []

for i in a:
    n = i
    s = 0

    while n > 0:
        s = s + (n % 10)
        n = n // 10

    if s % 3 == 0 and s > 5:
        b.append(i)

print(b)
```

```python
# Q3: Find the count of elements that are equal to the average (if any).

# sol:
a = (10, 20, 30, 20, 20)

s = 0
for i in a:
    s = s + i

avg = s / len(a)

c = 0
for i in a:
    if i == avg:
        c = c + 1

print(c)
```

```python
# Q4: Create a new set containing elements whose square plus the number is divisible by 6.

# sol:
a = {2, 3, 4, 5, 6}
b = set()

for i in a:
    if (i*i + i) % 6 == 0:
        b.add(i)

print(b)
```

```python
# Q5: Create a function that takes a list and returns a list of numbers whose first digit is greater than last digit.

# sol: 
def fun(a):
    b = []

    for i in a:
        last = i % 10
        n = i

        while n >= 10:
            n = n // 10

        first = n

        if first > last:
            b.append(i)

    return b

print(fun([23, 41, 56, 72, 91]))
```
