```python
# Q1: Count how many uppercase vowels are in a string.

# sol:
a = "ApplE OrangE"
c = 0

for i in a:
    if i in "AEIOU":
        c = c + 1

print(c)
```

```python
# Q2: From a list of numbers, create a new list of numbers that are smaller than the average of the list.

#sol:
a = [10, 20, 30, 40]

s = 0
for i in a:
    s = s + i

avg = s / len(a)

b = []
for i in a:
    if i < avg:
        b.append(i)

print(b)
```

```python
# Q3: Create a new dictionary with only the items whose values are multiples of 5.

# sol:
a = {"a": 10, "b": 12, "c": 20, "d": 7}
b = {}

for i in a:
    if a[i] % 5 == 0:
        b[i] = a[i]

print(b)
```

```python
# Q4: Convert a tuple into a set and:
#     print each element using a loop.

# sol:
a = (1, 2, 2, 3, 4)
b = set(a)

for i in b:
    print(i)
```

```python
# Q5: Create a function that:
#     takes a number
#     returns the sum of its digits using while loop.

# sol:
def sum_digits(n):
    s = 0
    while n > 0:
        s = s + (n % 10)
        n = n // 10
    return s

print(sum_digits(123))
```
