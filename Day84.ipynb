```python
# Q1: Check if number of digits greater than 5 is equal to number of uppercase consonants.

# sol:
a = "is c++ worlds 3rd hardest language or 5th one?"

d = 0
c = 0

for i in a:
    if i.isdigit():
        if int(i) > 5:
            d = d + 1
    elif i.isalpha():
        if i.isupper():
            if i not in "AEIOU":
                c = c + 1

if d == c:
    print("equal")
else:
    print("not equal")
```

```python
# Q2: Create a new list of elements where value is greater than square of index.

# sol:
a = [2, 5, 10, 3, 20]
b = []

for i in range(len(a)):
    if a[i] > i * i:
        b.append(a[i])

print(b)
```

```python
# Q3: Count elements where difference with previous element is between 2 and 5.

# sol:
a = (3, 6, 10, 13, 15)
c = 0

for i in range(1, len(a)):
    diff = a[i] - a[i - 1]

    if diff >= 2 and diff <= 5:
        c = c + 1

print(c)
```

```python
# Q4: Create a new set of elements whose cube + value is divisible by 5.

# sol:
a = {2, 3, 4, 5, 6}
b = set()

for i in a:
    if (i * i * i + i) % 5 == 0:
        b.add(i)

print(b)
```

```python
# Q5: Create a function that takes a list and returns numbers whose sum of digits is greater than 10.

# sol:
def fun(a):
    b = []

    for i in a:
        temp = i
        s = 0

        while temp > 0:
            s = s + (temp % 10)
            temp = temp // 10

        if s > 10:
            b.append(i)

    return b

print(fun([29, 56, 123, 41, 88]))
```
