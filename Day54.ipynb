```python
# Q1: Count how many characters in a string are either uppercase vowels or digits greater than 5.

# sol:
a = "PyThOn7 CodE89"
c = 0

for i in a:
    if i in "AEIOU" or (i.isdigit() and int(i) > 5):
        c = c + 1

print(c)
```

```python
# Q2: Create a new list containing numbers whose product of digits is greater than 10.

# sol:
a = [23, 14, 56, 11, 39]
b = []

for i in a:
    n = i
    p = 1

    while n > 0:
        p = p * (n % 10)
        n = n // 10

    if p > 10:
        b.append(i)

print(b)
```

```python
# Q3: Find the second smallest element in a tuple.

# sol:
a = (15, 7, 22, 5, 18)

m1 = a[0]
m2 = a[0]

for i in a:
    if i < m1:
        m2 = m1
        m1 = i
    elif i < m2 and i != m1:
        m2 = i

print(m2)
```

```python
# Q4: Create a new dictionary containing items whose values are multiples of 3 but not multiples of 2.

# sol:
a = {"a": 6, "b": 9, "c": 10, "d": 15}
b = {}

for i in a:
    if a[i] % 3 == 0 and a[i] % 2 != 0:
        b[i] = a[i]

print(b)
```

```python
# Q5: Create a function that takes a list and returns the sum of numbers whose square is odd.

# sol:
def total(a):
    s = 0

    for i in a:
        if (i * i) % 2 != 0:
            s = s + i

    return s

print(total([2, 3, 4, 5, 6]))
```
