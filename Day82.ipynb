```python
# Q1: Check if number of consonants is equal to number of digits.

# sol:
a = "Hello WORLd is 24x BEtter"

c = 0
d = 0

for i in a:
    if i.isalpha():
        if i not in "aeiou":
            c = c + 1
    elif i.isdigit():
        d = d + 1

if c == d:
    print("equal")
else:
    print("not equal")
```

```python
# Q2: Create a new list of elements whose value is greater than sum of its index and 2.

# sol:
a = [2, 6, 4, 9, 5]
b = []

for i in range(len(a)):
    if a[i] > i + 2:
        b.append(a[i])

print(b)
```

```python
# Q3: Count elements that are greater than previous and less than next.

# sol:
a = (3, 7, 5, 9, 6)
c = 0

for i in range(1, len(a) - 1):
    if a[i] > a[i - 1] and a[i] < a[i + 1]:
        c = c + 1

print(c)
```

```python
# Q4: Create a new set of elements whose cube is less than 200.

# sol:
a = {2, 3, 5, 6, 4}
b = set()

for i in a:
    if i * i * i < 200:
        b.add(i)

print(b)
```

```python
# Q5: Create a function that takes a list and returns numbers whose sum of digits is even .

# sol:
def fun(a):
    b = []

    for i in a:
        n = i
        s = 0

        while n > 0:
            s = s + (n % 10)
            n = n // 10

        if s % 2 == 0:
            b.append(i)

    return b

print(fun([12, 23, 34, 45, 56]))
```
