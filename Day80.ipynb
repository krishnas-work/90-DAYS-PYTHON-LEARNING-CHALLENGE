```python
# Q1: Check if a string has more vowels than consonants.

# sol:
a = "education"

v = 0
c = 0

for i in a:
    if i in "aeiou":
        v = v + 1
    elif i.isalpha():
        c = c + 1

if v > c:
    print("more vowels")
else:
    print("more consonants or equal")
```

```python
# Q2: Replace all numbers less than 5 with 0.

# sol:
a = [2, 7, 3, 9, 4]

for i in range(len(a)):
    if a[i] < 5:
        a[i] = 0

print(a)
```

```python
# Q3: Find the difference between maximum and minimum element.

# sol:
a = (5, 12, 3, 9, 7)

mx = a[0]
mn = a[0]

for i in a:
    if i > mx:
        mx = i
    if i < mn:
        mn = i

print(mx - mn)
```

```python
# Q4: Find the common elements between two sets.

# sol:
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}

c = set()

for i in a:
    if i in b:
        c.add(i)

print(c)
```

```python
# Q5: Create a function that takes a number and prints numbers from 1 to that number.

# sol:
def fun(a):
    i = 1

    while i <= a:
        print(i)
        i = i + 1

fun(5)
```
