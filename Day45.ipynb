```python
# Q1: Count how many characters in a string are either uppercase letters or digits.

# sol:
a = "PyTh0n123 Code"
c = 0

for i in a:
    if i.isupper() or i.isdigit():
        c = c + 1

print(c)
```

```python
# Q2: Create a new list containing numbers whose double is less than 25.

# sol:
a = [5, 8, 12, 15, 3]
b = []

for i in a:
    if i * 2 < 25:
        b.append(i)

print(b)
```

```python
# Q3: Count how many elements in a tuple are greater than the first element.

# sol:
a = (10, 5, 14, 8, 20)
c = 0
first = a[0]

for i in a:
    if i > first:
        c = c + 1

print(c)
```

```python
# Q4: Create a list of keys whose values are even numbers.

# sol:
a = {"a": 11, "b": 20, "c": 15, "d": 8}
b = []

for i in a:
    if a[i] % 2 == 0:
        b.append(i)

print(b)
```

```python
# Q5: Create a function that takes a tuple and returns the sum of elements that are less than 10.

# sol:
def total(a):
    s = 0

    for i in a:
        if i < 10:
            s = s + i

    return s

print(total((5, 12, 7, 3, 15)))
```
