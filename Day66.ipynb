```python
# Q1: Count how many characters in a string are alphabet letters and not spaces.

# sol:
a = "Python is Easy"
c = 0

for i in a:
    if i.isalpha():
        c = c + 1

print(c)
```

```python
# Q2: Create a new list containing numbers that are divisible by 2 or 5.

# sol:
a = [3, 4, 5, 7, 10, 12]
b = []

for i in a:
    if i % 2 == 0 or i % 5 == 0:
        b.append(i)

print(b)
```

```python
# Q3: Count how many elements are equal to 10.

# sol:
a = (10, 5, 10, 7, 3, 10)
c = 0

for i in a:
    if i == 10:
        c = c + 1

print(c)
```

```python
# Q4: Create a new set containing elements that are greater than or equal to 3.

# sol:
a = {1, 2, 3, 4, 5}
b = set()

for i in a:
    if i >= 3:
        b.add(i)

print(b)
```

```python
# Q5: Create a function that takes a dictionary and returns the sum of values less than 15.

# sol:
def fun(a):
    s = 0

    for i in a:
        if a[i] < 15:
            s = s + a[i]

    return s

print(fun({"a": 10, "b": 20, "c": 5, "d": 12}))
```
