```python
# Q1: Count how many times the letter 'a' appears in a string.

# sol: 
a = "banana"
c = 0

for i in a:
    if i == "a":
        c = c + 1

print(c)
```

```python
# Q2: From a list of numbers, create a new list of only positive numbers.

# sol:
a = [-3, 5, -1, 7, 0, 4]
b = []

for i in a:
    if i > 0:
        b.append(i)

print(b)
```

```python
# Q3: Find the total sum of all values in a dictionary.

# sol:
a = {"x": 10, "y": 20, "z": 30}
s = 0

for i in a:
    s = s + a[i]

print(s)
```

```python
# Q4: From a tuple, create a list of elements that are not equal to 0.

# sol:
a = (0, 5, 0, 7, 3)
b = []
for i in a:
    if i != 0:
        b.append(i)

print(b)
```

```python
# Q5: Create a function that:
#     takes a list
#     returns how many elements are greater than 10.

# sol:
def count(a):
    c = 0
    for i in a:
        if i > 10:
            c = c + 1
    return c

print(count([5, 12, 20, 3, 15]))
```
