```python
# Q1: Count how many characters in a string are uppercase consonants OR digits greater than 4.

# sol:
a = "HElloji567 CODE2"
c = 0

for i in a:
    if i.isalpha():
        if i.isupper():
            if i not in "AEIOU":
                c = c + 1
    elif i.isdigit():
        if int(i) > 4:
            c = c + 1

print(c)
```

```python
# Q2: Create a new list containing numbers that are less than 9 and not equal to 3.

# sol:
a = [2, 3, 7, 9, 1, 5]
b = []

for i in a:
    if i < 9:
        if i != 3:
            b.append(i)

print(b)
```

```python
# Q3: Count how many elements are greater than 6 and even.

# sol:
a = (4, 8, 3, 10, 5, 12)
c = 0

for i in a:
    if i > 6:
        if i % 2 == 0:
            c = c + 1

print(c)
```

```python
# Q4: Create a new set containing elements whose value − 2 is less than 5.

# sol:
a = {3, 5, 7, 2, 6}
b = set()

for i in a:
    if i - 2 < 5:
        b.add(i)

print(b)
```

```python
# Q5: Create a function that takes a list and returns the sum of numbers divisible by 3.

# sol:
def fun(a):
    s = 0

    for i in a:
        if i % 3 == 0:
            s = s + i

    return s

print(fun([3, 5, 6, 7, 9]))
```
