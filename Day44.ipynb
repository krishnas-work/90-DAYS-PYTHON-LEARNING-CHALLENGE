```python
# Q1: Find how many characters in a string are either vowels or digits.

# sol:
a = "pyth0n3 pr0gram"
c = 0

for i in a:
    if i in "aeiou" or i.isdigit():
        c = c + 1

print(c)
```

```python
# Q2: Create a new list containing numbers whose difference with the last element is less than 5.

# sol:
a = [12, 9, 15, 20, 18]
b = []

last = a[-1]

for i in a:
    if abs(i - last) < 5:
        b.append(i)

print(b)
```

```python
# Q3: Find the sum of elements that are multiples of 3 in a tuple.

# sol:
a = (6, 10, 9, 7, 12)
s = 0

for i in a:
    if i % 3 == 0:
        s = s + i

print(s)
```

```python
# Q4: Create a new set containing elements whose double is greater than 15.

# sol:
a = {4, 6, 8, 10}
b = set()

for i in a:
    if i * 2 > 15:
        b.add(i)

print(b)
```

```python
# Q5: Create a function that takes a list and returns the count of numbers greater than 10.

# sol:
def count_num(a):
    c = 0

    for i in a:
        if i > 10:
            c = c + 1

    return c

print(count_num([5, 12, 18, 7, 20]))
```
