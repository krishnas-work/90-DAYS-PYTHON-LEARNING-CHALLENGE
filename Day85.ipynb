```python
# Q1: Check if number of lowercase consonants is greater than digits that are odd.

# sol:
a = "You can CALL ME x 124"

c = 0
d = 0

for i in a:
    if i.isalpha():
        if i.islower():
            if i not in "aeiou":
                c = c + 1
    elif i.isdigit():
        if int(i) % 2 != 0:
            d = d + 1

if c > d:
    print("yes")
else:
    print("no")
```

```python
# Q2: Create a new list of elements where value is less than (index + 3) squared.

# sol:
a = [2, 5, 10, 4, 20]
b = []

for i in range(len(a)):
    if a[i] < (i + 3) * (i + 3):
        b.append(a[i])

print(b)
```

```python
# Q3: Count elements where it is greater than previous and difference with next is less than 4.

# sol:
a = (3, 7, 6, 10, 8)
c = 0

for i in range(1, len(a) - 1):
    if a[i] > a[i - 1]:
        if (a[i + 1] - a[i]) < 4:
            c = c + 1

print(c)
```

```python
# Q4: Create a new set of elements whose square is divisible by 4 and value is less than 10.

# sol:
a = {2, 3, 4, 5, 6, 7}
b = set()

for i in a:
    if (i * i) % 4 == 0:
        if i < 10:
            b.add(i)

print(b)
```

```python
# Q5: Create a function that takes a list and  returns numbers whose sum of digits is less than the number itself.

# sol:
def fun(a):
    b = []

    for i in a:
        temp = i
        s = 0

        while temp > 0:
            s = s + (temp % 10)
            temp = temp // 10

        if s < i:
            b.append(i)

    return b

print(fun([12, 25, 39, 44, 51]))
```
