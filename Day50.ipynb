```python
# Q1: Count how many characters in a string are letters that appear more than once.

# sol:
a = "programming"
c = 0

for i in a:
    if a.count(i) > 1:
        c = c + 1

print(c)
```

```python
# Q2: Create a new list containing numbers whose sum of first and last digit is greater than 10.

# sol:
a = [23, 45, 89, 12, 67]
b = []

for i in a:
    x = i % 10
    y = i
    while y >= 10:
        y = y // 10
    if x + y > 10:
        b.append(i)

print(b)
```

```python
# Q3: Count how many elements in a tuple are divisible by both 2 and 5.

# Sol:
a = (10, 15, 20, 22, 25, 30)
c = 0

for i in a:
    if i % 2 == 0 and i % 5 == 0:
        c = c + 1

print(c)
```

```python
# Q4: Create a new set containing elements whose cube is less than 200.

# sol: 
a = {2, 3, 5, 6}
b = set()

for i in a:
    if i**3 < 200:
        b.add(i)

print(b)
```

```python
# Q5: Create a function that takes a dictionary and returns the product of all even values.

# sol:
def product(a):
    p = 1

    for i in a:
        if a[i] % 2 == 0:
            p = p * a[i]

    return p

print(product({"a": 2, "b": 5, "c": 4, "d": 3}))
```
